```python
#all importing is done here

import numpy as np
import scipy.io as sio
import pandas as pd
import matplotlib.pyplot as plt
import itertools as it
from scipy import stats
```

```python
wine_data = sio.loadmat('wine.mat')
data = wine_data['X']
labels = wine_data['attributeNames'][0]

for i in range(len(labels)):
  labels[i] = labels[i][0]

df = pd.DataFrame(data, columns=labels)

#With outliers
fig1, axs1 = plt.subplots(3, 4, constrained_layout=True)
for i in range(3):
  for j in range(4):
    axs1[i, j].boxplot(stats.zscore(df[labels[4*i+j]]))
    axs1[i,j].set_title(labels[4*i+j])
fig1.suptitle("Box plots of attributes")
plt.show()

fig2, axs2 = plt.subplots(3, 4, constrained_layout=True)
for i in range(3):
  for j in range(4):
    axs2[i, j].hist(df[labels[4*i+j]])
    axs2[i,j].set_title(labels[4*i+j])
fig2.suptitle("Histograms of attributes")
plt.show()
```

```python
#Time to remove the outliers. I made the observation that the perentage can never be higher than 100% 

df = df.query("Alcohol <= 100").query("Density <= 10")
df = df.loc[df['Volatile acidity'] <= 20]

#Without outliers
fig1, axs1 = plt.subplots(3, 4, constrained_layout=True)
for i in range(3):
  for j in range(4):
    axs1[i, j].boxplot(stats.zscore(df[labels[4*i+j]]))
    axs1[i,j].set_title(labels[4*i+j])
fig1.suptitle("Box plots of attributes")
plt.show()

fig2, axs2 = plt.subplots(3, 4, constrained_layout=True)
for i in range(3):
  for j in range(4):
    axs2[i, j].hist(df[labels[4*i+j]])
    axs2[i,j].set_title(labels[4*i+j])
fig2.suptitle("Histograms of attributes")
plt.show()
```

In the first set of box plots and histograms it is instantly clear that there are some huge outliers in the Volatile acidity, Density and Alcohol data. In the corresponding box plots, the boxes are barely visible due to the extremity of the outliers. In the histograms almost all of the data is thrown in the same bin due to these outliers.

When leaving out the biggest outliers we see a huge difference. The data is more spread over the different bins in the histograms and the boxes of the box plots are actually visible. However except from the Density there are still outliers apperent.

+ *Keep in mind that numpy.linalg.svd() returns the transposed **V<sup>T</sup>** matrix as output.*

+ *To reconstruct the data from projection you can use the following formula: **W** = **Z**&ast;**V**[:,:4]<sup>T</sup> + ****. *

+ *It is advisable to make a for-loop to generate the 2D scatter plots, this saves a lot of time. It is an important skill to master if you want to easily modify your work later on, for example when correcting mistakes, or when you want to modify each plot in the same manner.* 


```python
## Example code:
#------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.io import loadmat
from numpy import reshape


# Load Matlab data file to python dict structure
mat_data = loadmat('zipdata.mat')

# Extract variables of interest
testdata = mat_data['testdata']
traindata = mat_data['traindata']
X = traindata[:,1:]
y = traindata[:,0]

for i in range(10):
  # Visualize the i'th digit as an image
  plt.subplot(121);
  I = reshape(X[i,:],(16,16))
  plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
  plt.title('Original image');

  
  mu = np.zeros(len(I))
  for j in range(len(mu)):
    mu[j] = np.mean(I[:, j])
  Y = I - mu
  decomp = np.linalg.svd(Y)
  V = decomp[2]
  Z = np.dot(Y, V[:, :4])
  W = np.dot(Z, V[:,:4].T) + mu
  
  plt.subplot(122);
  J = reshape(W,(16,16))
  plt.imshow(J, extent=(0,16,0,16), cmap=cm.gray_r);
  plt.title('Image after PCA');
  plt.show()
  
  #------------------------------------------------
```

After PCA most of the images are still somewhat recognizable.

> v. (0.3pts) Plot the distribution of the population and the distributions of both sample means using histograms. What happens to the shape of the sample means distribution as the sample size (N*) increases?

```python
def get_sample_means(data, N):
  comb = list(it.combinations(data, N)) 
  lst = []
  for i in comb:
    lst.append(np.mean(i))
  return lst

    
fig, axs = plt.subplots(2, 3, constrained_layout=True)
for i in range(2):
  for j in range(3):
    axs[i, j].hist(get_sample_means(data, 3*i+j+1))
    axs[i, j].set_title("N = %d" % (3*i+j+1))
    axs[i, j].set(xlabel='Frequency', ylabel='Limbs')
fig.suptitle("Sample mean distributions")
plt.show()
```

Due to the central limi theorem we expect to see something resembling a normal distribution. However the number of samples we take is to small to really find something similar. However we obtain the same number of samples for N = 1 and N =5, for N = 2 and N = 4. Those two couples of N seem to give similar histograms. The biggest ammount of samples is found for N=3, the corresponding histogram looks most like a normal distribution.
