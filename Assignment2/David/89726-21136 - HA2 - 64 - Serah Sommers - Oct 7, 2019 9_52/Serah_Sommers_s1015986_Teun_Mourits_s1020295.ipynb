```python
## Example code:
#------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.io import loadmat
from numpy import reshape
import itertools


# Load Matlab data file to python dict structure
mat_data = loadmat('./Data/zipdata.mat')

# Extract variables of interest
testdata = mat_data['testdata']
traindata = mat_data['traindata']
X = traindata[:,1:]
y = traindata[:,0]


t = y<2
filtereddata = X[t, :]

#plotting 10 digits as images
for n in range(10):
    plt.subplot(1,1,1);
    I = reshape(filtereddata[n,:],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image');
    plt.show()

#computing PCA
q = filtereddata.mean(0)
Y = filtereddata.copy()
for x in range(0, len(q)):
    Y[x]-q[x]  


SVD = np.linalg.svd(Y)
V = SVD[2].T
Z = Y.dot(V[:,:4])
W = Z.dot(V[:,:4].T)+q

for m in range(10):
    plt.subplot(1,1,1)
    I = reshape(W[m,:],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('W digits as an image');
    plt.show()
    
    
#The images of the original data is way more clear than the images of the data in W. 
#In the original data you can clearly recognize what digits are 0's and 1's. In the data from W, this is not so clear. 
array=[]
for x in itertools.combinations(Z, 2):
    array.append(x)


fig, axs = plt.subplots(4,4, figsize = (15,15))
fig.subplots_adjust(hspace = 0.5, wspace=0.001)
for i in range(4):
    for j in range(4):
        axs[i,j].plot(Z[:,i],Z[:,j],'.')
   

   
#3D scatterplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
PC1 =Z[0][0:]
PC2 =Z[1][0:]
PC3 =Z[2][0:]

ax.scatter(PC1, PC2, PC3, c='r', marker='o')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.show()

#We can conclude that normal 2D scatterplots have contain more data, that is more points of the PC's that are shown 
#and a scatterplot for every possible combination of vectors. 
#The 3D scatterplot has axes for every PC, so it is more complex and harder to understand, but more compact in showing data.

```
