2. Create the data matrix *X* and the class index vector *y* from the data. Remove
the digits with the class index 2-9 from the data, so only digits belonging to

3. Visualize the first 10 digits as images. (take a look at the example code)
Next, compute the principal components (PCA) of the data matrix. Now, using the PCA model, create a new data matrix $Z$ by projecting $X$ onto the space spanned by the loadings $V$. The new data matrix should have 4 attributes corresponding to PC1-PC4.  Use subplotting to show the digits and their reconstructed counterparts in an orderly manner.
4. Reconstruct the initial data using PC1-PC4 into a new matrix called $W$. Visualize the first 10 digits as images for the reconstructed data and compare them with images for the original data.

> **Hints:**
+ *The below example code can help you visualize digits as images.*
+ *See Assignment 1 if you can not recall how to compute a PCA.*
+ *Keep in mind that numpy.linalg.svd() returns the transposed **V<sup>T</sup>** matrix as output.*
+ *You can use **Z** = **Y** $*$ **V**[:,:4] to project the data onto the first four PCs. Don't forget that the $*$ operator does not perform matrix multiplication for numpy arrays!*
+ *To reconstruct the data from projection you can use the following formula: **W** = **Z**&ast;**V**[:,:4]<sup>T</sup> + ****. *
+ *You can take a look at the example_figure.ipynb notebook to see how you can easily plot multiple classes and color them correspondingly.* 
+ *It is advisable to make a for-loop to generate the 2D scatter plots, this saves a lot of time. It is an important skill to master if you want to easily modify your work later on, for example when correcting mistakes, or when you want to modify each plot in the same manner.* 


```python
## Example code:
#------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.io import loadmat
from numpy import reshape

# Index of the digit to display
i = 0

# Load Matlab data file to python dict structure
mat_data = loadmat('./Data/zipdata.mat')

# Extract variables of interest
testdata = mat_data['testdata']
traindata = mat_data['traindata']
X = traindata[:,1:]
y = traindata[:,0]

# Visualize the i'th digit as an image
plt.subplot(1,1,1);
I = reshape(X[i,:],(16,16))
plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
plt.title('Digit as an image');
plt.show()
#------------------------------------------------

##Answer to question 2.2.1 - part 1

#Remove the digits with the class index 2-9 from the data, since indices go from 0-9 everything higher than 1 can be removed
to_remove = np.where(y > 1)
X = np.delete(X, to_remove, axis = 0)
y = np.delete(y, to_remove)

#first 10 digits as images: 
print('First 10 digits: ')
i = 0
plt.subplot(1,1,1);
for i in range(5): 
    for j in range(2): 
        I = reshape(X[i,:],(16,16))
        plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
        plt.title('Digit as an image #'+str(i+1));
        plt.show()

#PCA
#mean center
Y = X - X.mean(axis=0) 
# Y = np.matrix(Y, dtype='float')

#perform svd
U,S,V = np.linalg.svd(Y)

#transpose of V
V = V.transpose()

#project data onto the PCs
PC1 = np.dot(X, V[:,0]) #PC1
PC2 = np.dot(X, V[:,1]) #PC2
PC3 = np.dot(X, V[:,2]) #PC3
PC4 = np.dot(X, V[:,3]) #PC4

Z = Y.dot(V[:,:4])
W = np.dot(Z, np.transpose(V[:, :4])) + X.mean(axis=0)


#Reconstructed stuff
print('Reconstructed digits: ')
i = 0
plt.subplot(1,1,1);
for i in range(5): 
    for j in range(2): 
        IW = reshape(W[i,:],(16,16))
        plt.imshow(IW, extent=(0,16,0,16), cmap=cm.gray_r);
        plt.title('Digit as an image #'+str(i+1));
        plt.show()
plt.show()


#Scatter plots
print('Possible combination projections')
fig,axs = plt.subplots(6, figsize=(20,30))
counter = 0
for i in range(4):
    for j in range(i+1,4):
        if(i != j):
            axs[counter].scatter(Z[:,i],Z[:,j])
            axs[counter].set_title('Scatterplot combination pc'+str(i+1)+' onto pc'+str(j+1))
            axs[counter].set_xlabel('PC'+str(i+1))
            axs[counter].set_ylabel('PC'+str(j+1))
            counter+=1
plt.show()

# 3d plot
from mpl_toolkits.mplot3d import Axes3D

#0 digits
class_1 = []
for i in range(len(y)):
    if y[i] == 0:
        class_1.append(Z[i])
class_1 = np.array(class_1)

#1 digits
class_2 = []
for i in range(len(y)):
    if y[i] == 1: 
        class_2.append(Z[i])
class_2 = np.array(class_2)
        
fig = plt.figure()
axs = Axes3D(fig)
axs.scatter(class_1[:,0], class_1[:,1], class_1[:,2], label = '0')
axs.scatter(class_2[:,0], class_2[:,1], class_2[:,2], label = '1')
axs.set_xlabel('PC1')
axs.set_ylabel('PC2')
axs.set_zlabel('PC3')
axs.legend()
plt.show()
```

```python
##Answer to question 2.3.2
from itertools import combinations
combinations_of_two_samples = np.array(list(combinations(alien_data,2)))
combinations_of_two_samples_with_mean = np.mean(combinations_of_two_samples, axis = 1)
print('two samples: ',combinations_of_two_samples)
print('two samples with mean: ', combinations_of_two_samples_with_mean)

combinations_of_four_samples = np.array(list(combinations(alien_data,4)))
combinations_of_four_samples_with_mean = np.mean(combinations_of_two_samples, axis = 1)
print('four samples: ',combinations_of_four_samples)
print('four samples with mean: ', combinations_of_four_samples_with_mean)
```
