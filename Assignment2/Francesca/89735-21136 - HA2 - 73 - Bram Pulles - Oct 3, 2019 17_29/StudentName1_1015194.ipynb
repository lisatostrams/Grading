```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
from scipy.stats import zscore
%matplotlib inline 

mat_contents = loadmat('Data/wine.mat')

X = mat_contents['X']
attributeNames = mat_contents['attributeNames']

newAttributeNames = []
for arr in attributeNames[0]:
    newAttributeNames.append(arr[0])   

attributeNames = newAttributeNames # Can this be done easier?

# Histograms
plt.figure(figsize=(30,10))

for i in range(0, len(attributeNames)):
    plt.subplot(1, len(attributeNames), i + 1)
    plt.hist(X[:,i])
    plt.title(attributeNames[i])

plt.show()
print("Histograms of the wine data.")

# Boxplots
plt.figure(figsize=(20,10))

plt.boxplot(zscore(X))
plt.xticks(range(1, len(attributeNames) + 1), attributeNames)

plt.show()
print("Normalized boxplots of the wine data.")

# Filter the outliners
X_filtered = [x for x in X if (x[1] < 20 and x[7] < 10 and x[10] < 200)]

# Boxplots without outliners
plt.figure(figsize=(20,10))

plt.boxplot(zscore(X_filtered))
plt.xticks(range(1, len(attributeNames) + 1), attributeNames)

plt.show()
print("Normalized boxplots of the wine data without outliners.")
```

> **Hints:**
+ *The below example code can help you visualize digits as images.*
+ *See Assignment 1 if you can not recall how to compute a PCA.*
+ *Keep in mind that numpy.linalg.svd() returns the transposed **V<sup>T</sup>** matrix as output.*
+ *You can use **Z** = **Y** $*$ **V**[:,:4] to project the data onto the first four PCs. Don't forget that the $*$ operator does not perform matrix multiplication for numpy arrays!*
+ *To reconstruct the data from projection you can use the following formula: **W** = **Z**&ast;**V**[:,:4]<sup>T</sup> + **μ**. *
+ *You can take a look at the example_figure.ipynb notebook to see how you can easily plot multiple classes and color them correspondingly.* 
+ *It is advisable to make a for-loop to generate the 2D scatter plots, this saves a lot of time. It is an important skill to master if you want to easily modify your work later on, for example when correcting mistakes, or when you want to modify each plot in the same manner.* 


```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from scipy.io import loadmat

mat_data  = loadmat('./Data/zipdata.mat')
traindata = mat_data['traindata']

# Filter the 0's and 1's from the traindata.
traindata_filtered = [x for x in traindata if (x[0] <= 1)]
traindata_filtered = np.array(traindata_filtered)

X = traindata_filtered[:,1:]
X_answers = traindata_filtered[:,0]

# Calculate the PCA.
mu      = np.mean(X,axis=0)
Y       = X-mu
U,sv,Vt = np.linalg.svd(Y)
V       = Vt.T
Z       = np.dot(Y, V[:,:4])

# Calculate the reconstructed dataset.
W = np.dot(Z, np.transpose(V[:,:4])) + mu

# Print all the digits.
fig, axs = plt.subplots(10, 3, figsize=(30,30))
for i in range(0, 10):
    # Visualize the i'th original digit as an image.
    I = np.reshape(X[i,:],(16,16))
    axs[i, 0].imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    axs[i, 0].set_title("Image " + str(i) + " in our original dataset.");
    # Visualize the i'th new digit as an image.
    I = np.reshape(Z[i,:],(2,2))
    axs[i, 1].imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    axs[i, 1].set_title("Image " + str(i) + " in our new dataset.");
    # Visualize the i'th reconstructed digit as an image.
    I = np.reshape(W[i,:],(16,16))
    axs[i, 2].imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    axs[i, 2].set_title("Image " + str(i) + " in our reconstructed dataset.");

plt.tight_layout(pad=0, w_pad=0, h_pad=0)

# Print the PCA's against each other.
fig, axs = plt.subplots(4, 4, figsize=(30,30))
colors = ['blue','orange']
for y in range(0, 4):
    for x in range(0, 4):
        if x != y:
            axs[y, x].scatter(Z[:,x][X_answers == 0], Z[:,y][X_answers == 0], color='blue')
            axs[y, x].scatter(Z[:,x][X_answers != 0], Z[:,y][X_answers != 0], color='red')
            axs[y, x].set_title("PCA " + str(y) + " on the y-axis and PCA " + str(x) + " on the x-axis.")
print("The digit 0 is shown in blue and the digit 1 is shown in red.")          

# Print three PCA's in a 3D scatter plot.
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Z[:,0][X_answers == 0], Z[:,1][X_answers == 0], Z[:,2][X_answers == 0], color='blue')
ax.scatter(Z[:,0][X_answers != 0], Z[:,1][X_answers != 0], Z[:,2][X_answers != 0], color='red')

ax.set_xlabel('PCA 0')
ax.set_ylabel('PCA 1')
ax.set_zlabel('PCA 2')
plt.show()
```

> ii. (0.3pts) List all possible samples of two aliens without replacement, and find each mean. Do the same with samples of four aliens.
+ *You can use the method itertools.combinations(v,n) to find all possible samples of a vector v taking n elements at a time.*

```python
from itertools import combinations

def it(n):
    print("Samples with " + str(n) + " aliens.")
    for i in combinations([2,3,6,8,11,18], n):
        print("Combination: " + str(i) + "\thas mean: " + str(np.ndarray.mean(np.array(i))))    

it(2)
it(4)
```

> iii. (0.3pts) Each of the means above is called a sample mean. Find the mean of all the sample means (denoted by $m_x$) and the standard
deviation of all the sample means (denoted by $\sigma_x$) for both
the *N=2* and *N=4* samples.

```python
def everything(n,f):
    combos = []
    for i in combinations([2,3,6,8,11,18], n):
        combos.append(f(np.array(i)))
    return f(np.array(combos))
        
print("Mean 2: " + str(everything(2,np.ndarray.mean)))
print("Standard deviation 2: " + str(everything(2,np.ndarray.std)))
print("Mean 4: " + str(everything(4,np.ndarray.mean)))
print("Standard deviation 4: " + str(everything(4,np.ndarray.std)))
```

> iv. Verify the Central Limit Theorem: (i) (0.1pts) compare the population
mean with the mean of both sample means; (ii) (0.2pts) compare the population
standard deviation divided by the square root of the sample size
with the standard deviation of both sample means (i.e., does
$\sigma_x \approx \sigma/\sqrt{N}$). BTW, a better approximation for
small population sizes is $\sigma_x = \sigma / \sqrt{N} \times
\sqrt{(M-N)/(M-1)}$ with *M = 6* the size of the original

```python
import math

print("n = 2")
print(str(np.ndarray.std(np.array([2,3,6,8,11,18])) / math.sqrt(2)) + " = " + str(everything(2,np.ndarray.std)))

print("n = 4")
print(str(np.ndarray.std(np.array([2,3,6,8,11,18])) / math.sqrt(4)) + " = " + str(everything(4,np.ndarray.std)))
```
