```python
import scipy.io
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

data = scipy.io.loadmat("Data/wine.mat")
x = data['X']
filtered_x = x
filtered_x = filtered_x[(filtered_x[ :,1]<20)]
filtered_x = filtered_x[((filtered_x[ :,7]> 0.5) & (filtered_x[ :,7]<1.5))]
filtered_x = filtered_x[((filtered_x[ :,10] >= 5) & (filtered_x[ :,10] <= 20))]

standardized_x_with_outliers = scipy.stats.zscore(x)
standardized_x = scipy.stats.zscore(filtered_x)

#boxplot
plt.figure(figsize=(5,5))
plt.boxplot(standardized_x_with_outliers)
plt.show()

#histogram
plt.figure(figsize=(5,5))
plt.hist(standardized_x_with_outliers)
plt.show()

# boxplot subplot
fig, ax = plt.subplots(3, 4, figsize = (15,15))
ax = ax.ravel()
for i in range(0,12):
    ax[i].boxplot(standardized_x_with_outliers[: , i])
    ax[i].set_title('Boxplot')
    
# histogram subplot 
fig, ax = plt.subplots(3, 4, figsize = (15,15))
ax = ax.ravel()
for i in range(0,12):
    ax[i].hist(x[: , i])
    ax[i].set_title('Histogram')

#boxplot without the outliers
fig, ax = plt.subplots(3, 4, figsize = (15,15))
ax = ax.ravel()
for i in range(0,12):
    ax[i].boxplot(standardized_x[: , i])

#histogram without the outliers 
fig, ax = plt.subplots(3, 4, figsize = (15,15))
ax = ax.ravel()
for i in range(0,12):
    ax[i].hist(filtered_x[: , i])
```

----
We can see that the histograms with outliers are only one bar, while the histograms without the outliers have multiple bars next to each other. 
Also the boxplots with outliers are way more stretched out, you cannot really see the points because the outliers are way bigger than the rest. The boxplots without outliers are more precise and it is clear to see where the dots are.

> **Hints:**
+ *The below example code can help you visualize digits as images.*

+ *Keep in mind that numpy.linalg.svd() returns the transposed **V<sup>T</sup>** matrix as output.*
+ *You can use **Z** = **Y** $*$ **V**[:,:4] to project the data onto the first four PCs. Don't forget that the $*$ operator does not perform matrix multiplication for numpy arrays!*

+ *It is advisable to make a for-loop to generate the 2D scatter plots, this saves a lot of time. It is an important skill to master if you want to easily modify your work later on, for example when correcting mistakes, or when you want to modify each plot in the same manner.* 


```python
## Example code:
#------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as panda
import math
from mpl_toolkits.mplot3d import Axes3D
from scipy.io import loadmat
from numpy import reshape

# Index of the digit to display
# i = 4

# Load Matlab data file to python dict structure
mat_data = loadmat('./Data/zipdata.mat')
# Extract variables of interest
testdata = mat_data['testdata']
traindata = mat_data['traindata']
X = traindata[:,1:]
y = traindata[:,0]

# 2
Y_booleans = y<2
Y_filtered = y[Y_booleans]
X_filtered = X[Y_booleans]

# 3
for i in range (0,10):
    plt.subplot(1,1,1);
    I = reshape(X_filtered[i,:],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image');
    plt.show()
    
# 1. calculate the mean
m = np.mean(X_filtered, axis=0)
# 2. center the data
Y = X_filtered - m
# 3. Compute the singular values and vectors
U,sv,Vt = np.linalg.svd(Y)
V = Vt.T
# 4. Calculate the lower dimension representation
D = np.dot(Y,V[:, 1])
C = np.dot(Y,V[:, 2])
B = np.dot(Y,V[:, 3])
A = np.dot(Y,V[:, 4])
K = [D, C, B, A]

U,sv,Vt = np.linalg.svd(Y)
V = Vt.T
Z = Y.dot(V[:,:4])
W = np.dot(Z, V[:,:4].T) - m
# print(W)

# 4
for i in range (0,10):
    plt.subplot(1,1,1);
    I = reshape(W[i,:],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image');
    plt.show()

# 5
plt.figure(figsize=(80,80))
plt.xlabel("PC", fontsize = 80)
plt.ylabel("PC", fontsize = 80)
plt.scatter(K[0], K[1], label="PC1 and PC2")
plt.scatter(K[1], K[0], label="PC2 and PC1")
plt.scatter(K[1], K[2], label="PC2 and PC3")
plt.scatter(K[2], K[1], label="PC3 and PC2")
plt.scatter(K[2], K[3], label="PC3 and PC4")
plt.scatter(K[3], K[2], label="PC4 and PC2")
plt.scatter(K[2], K[0], label="PC3 and PC1")
plt.scatter(K[0], K[2], label="PC1 and PC3")
plt.scatter(K[3], K[1], label="PC4 and PC2")
plt.scatter(K[1], K[3], label="PC2 and PC4")
plt.scatter(K[0], K[3], label="PC1 and PC4")
plt.scatter(K[3], K[0], label="PC4 and PC1")
plt.legend(fontsize = 80)

# 6
fig = plt.figure(figsize=(30,30))
ax = Axes3D(fig)
ax.set_xlabel("PC 1", fontsize = 30)
ax.set_ylabel("PC 2", fontsize = 30)
ax.set_zlabel("PC 3", fontsize = 30)
ax.scatter(K[0], K[1], K[2], label="PC1, PC2 and PC3")
ax.legend(fontsize = 40)
#------------------------------------------------
```

----
We can conclude from the scatterplots that the PC's separate the data in blobs. The dots are close to each other on a certain place and all the PC's together form the most important information, which is in the middle of the blobs.
