> iii. (0.3pts) Each of the means above is called a sample mean. Find the mean of all the sample means (denoted by $m_x$) and the standard
deviation of all the sample means (denoted by $\sigma_x$) for both
the *N=2* and *N=4* samples.

```python
combination_2 = itertools.combinations(limbs, 2)
combination_4 = itertools.combinations(limbs, 4)

means_2 = np.array([])
for comb in combination_2:
    means_2 = np.append(means_2, np.mean(comb))
    
means_4 = np.array([])
for element in combination_4:
    means_4 = np.append(means_4, np.mean(element))

m_2 = np.mean(means_2)
_2 = np.std(means_2)
m_4 = np.mean(means_4)
_4 = np.std(means_4)

print('n = 2 \n Mean: ' + str(m_2) + '\n Std: ' + str(_2))
print('n = 4 \n Mean: ' + str(m_4) + '\n Std: ' + str(_4))
```
