```python
##Answer to question 2.2.1
from numpy import reshape
import matplotlib.cm as cm
zipdata=loadmat('zipdata.mat')
traindata = zipdata['traindata']
Y=[]
for i in range(0,len(traindata)):
    Y.append(traindata[i][0])
#print(Y)
X=[ row[1:] for row in traindata ]
X1=[]
#print(X)
for i in range(0,len(Y)):
    if Y[i]<2:
        X1.append(X[i])
for i in range(0,10):
    plt.subplot(1,1,1);
    I = reshape(X1[i],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image (original)');
    plt.show()
    
#PCA
#mean centering the data
mu = np.mean(X1,axis=0)
#len(X1)#2199
one = np.ones(2199)
one.shape = (2199,1)
Y = X1-(one*mu)
#SVD
U,S,V = np.linalg.svd(Y)
V_transposed = V.transpose()
#create the new data matrix Xpca
Xpca = []
Xpca.append(Y*V[:,0])
Xpca.append(Y*V[:,1])
Xpca.append(Y*V[:,2])
Xpca.append(Y*V[:,3])
#data reconstruction
reconstruted_data = Xpca[0]+Xpca[1]+Xpca[2]+Xpca[3]
#visualize the first 10 digits of the reconstructed data
for i in range(0,10):
    plt.subplot(1,1,1);
    I = reshape(reconstruted_data[i],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image (reconstructed)');
    plt.show()
#scatter plots of two pca's
plt.scatter(Xpca[0],Xpca[1])
plt.title('PCA1 against PCA2 ')
plt.ylabel('PCA2')
plt.xlabel('PCA1')
plt.show()
plt.scatter(Xpca[0],Xpca[2])
plt.title('PCA1 against PCA3 ')
plt.ylabel('PCA3')
plt.xlabel('PCA1')
plt.show()
plt.scatter(Xpca[0],Xpca[3])
plt.title('PCA1 against PCA4 ')
plt.ylabel('PCA4')
plt.xlabel('PCA1')
plt.show()
plt.scatter(Xpca[1],Xpca[2])
plt.title('PCA2 against PCA3 ')
plt.ylabel('PCA3')
plt.xlabel('PCA2')
plt.show()
plt.scatter(Xpca[1],Xpca[3])
plt.title('PCA2 against PCA4 ')
plt.ylabel('PCA4')
plt.xlabel('PCA2')
plt.show()
plt.scatter(Xpca[2],Xpca[3])
plt.title('PCA3 against PCA4 ')
plt.ylabel('PCA4')
plt.xlabel('PCA3')
plt.show()
```
