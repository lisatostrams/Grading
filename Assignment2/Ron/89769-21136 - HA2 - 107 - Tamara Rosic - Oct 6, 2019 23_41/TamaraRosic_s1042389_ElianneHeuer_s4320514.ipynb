Why do we need to standardize the data after removing the outliers? Give the -statistical- reason, not just the practical reason. 

----
We need to standardize data when we're looking at relationships in data. Outliers skew the data in a positive or negative direction, so after removing them we need to standardise the data so we are able to scale it consistently and  we can compare it more easily. In statistics, outliers will drastically change the result of calculations measuring the spread such as the mean, standard deviation and variance. The reason is that these calculations use each data point, so outliers will affect the result.

+ *You can use the method itertools.combinations(v,n) to find all possible samples of a vector v taking n elements at a time.*

```python
import itertools as it
listMeans2 = []
listMeans4 = []

list2 = list(it.combinations(arr, 2))
print(list2)
for i in list2:
     listMeans2.append(sum(i) / 2)
        
print(listMeans2)

list4 = list(it.combinations(arr, 4))
print(list4)
for i in list4:
     listMeans4.append(sum(i) / 4)
        
print(listMeans4)
```
