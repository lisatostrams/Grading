> **Hints:**
+ *The below example code can help you visualize digits as images.*

+ *Keep in mind that numpy.linalg.svd() returns the transposed **V<sup>T</sup>** matrix as output.*
+ *You can use **Z** = **Y** $*$ **V**[:,:4] to project the data onto the first four PCs. Don't forget that the $*$ operator does not perform matrix multiplication for numpy arrays!*
+ *To reconstruct the data from projection you can use the following formula: **W** = **Z**&ast;**V**[:,:4]<sup>T</sup> + **μ**. *

+ *It is advisable to make a for-loop to generate the 2D scatter plots, this saves a lot of time. It is an important skill to master if you want to easily modify your work later on, for example when correcting mistakes, or when you want to modify each plot in the same manner.* 


```python
## Example code:
#------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy
from scipy.io import loadmat
from numpy import reshape
from mpl_toolkits.mplot3d import Axes3D

# Load Matlab data file to python dict structure
mat_data = loadmat('./Data/zipdata.mat')

# Extract variables of interest
testdata = mat_data['testdata']
traindata = mat_data['traindata']
X = traindata[:,1:]
y = traindata[:,0]

# Visualize the i'th digit as an image
for i in range(10):
    plt.figure()
    I = reshape(X[i,:],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image');
    plt.show()
#------------------------------------------------

m = X.mean(axis=0)
Y = X - m

U, S, Vt = numpy.linalg.svd(Y)
V = numpy.transpose(Vt)

Z = numpy.dot(Y, V[:,:4])
W = numpy.dot(Z, numpy.matrix.transpose(V[:, :4])) + numpy.mean(Z)

#reconstructed data visualisation
for i in range(10):
    plt.figure()
    I = reshape(W[i,:], (16,16))
    plt.imshow(I,extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image')
    plt.show()
    

fig, axs = plt.subplots(4,3)
axs[0,0].scatter(Z[:,0],Z[:,1])
axs[0,0].set_title("PC1 vs PC2")
axs[0,1].scatter(Z[:,0],Z[:,2])
axs[0,1].set_title("PC1 vs PC3")
axs[0,2].scatter(Z[:,0],Z[:,3])
axs[0,2].set_title("PC1 vs PC4")
axs[1,0].scatter(Z[:,1],Z[:,0])
axs[1,0].set_title("PC2 vs PC1")
axs[1,1].scatter(Z[:,1],Z[:,2])
axs[1,1].set_title("PC2 vs PC3")
axs[1,2].scatter(Z[:,1],Z[:,3])
axs[1,2].set_title("PC2 vs PC4")
axs[2,0].scatter(Z[:,2],Z[:,0])
axs[2,0].set_title("PC3 vs PC1")
axs[2,1].scatter(Z[:,2],Z[:,1])
axs[2,1].set_title("PC3 vs PC2")
axs[2,2].scatter(Z[:,2],Z[:,3])
axs[2,2].set_title("PC3 vs PC4")
axs[3,0].scatter(Z[:,3],Z[:,0])
axs[3,0].set_title("PC4 vs PC1")
axs[3,1].scatter(Z[:,3],Z[:,1])
axs[3,1].set_title("PC4 vs PC2")
axs[3,2].scatter(Z[:,3],Z[:,2])
axs[3,2].set_title("PC4 vs PC3")



fig1 = plt.figure()
ax=Axes3D(fig1)
ax.scatter(Z[:,0], Z[:,1], Z[:,2])
plt.xlabel('PC1')
plt.ylabel('PC2')
ax.set_zlabel('PC3')
plt.title('3D scatterplot PC1, PC2, PC3')
```

> 2.3.1 (0.3pts) A study of a very limited population of Aliens reveals the following number of body appendages (limbs):

i. Find the mean $m$ and the standard deviation $\sigma$ of this population.
+ *You can use the methods numpy.ndarray.mean() and numpy.ndarray.std() to calculate the mean and standard deviation.*

```python
import numpy as np

x = np.array([2,3,6,8,11,18])
print("Mean of x: ", np.ndarray.mean(x))
print("Standard deviation of x: ", np.ndarray.std(x))
```

> ii. (0.3pts) List all possible samples of two aliens without replacement, and find each mean. Do the same with samples of four aliens.
+ *You can use the method itertools.combinations(v,n) to find all possible samples of a vector v taking n elements at a time.*

```python
import itertools as it

list1 =  it.combinations(x, 2)
two_aliens = []
for i in list1:
    two_aliens.append(i)  
print("Combinations of two aliens without replacement: ", two_aliens)

means_ta = []
for i in two_aliens:  
    means_ta.append(np.ndarray.mean(np.array(i)))
print("Means of combinations of two aliens: ", means_ta)

list2 = it.combinations(x, 4)

four_aliens = []
for i in list2:
    four_aliens.append(i)
print("Combinations of four aliens without replacement: ", four_aliens)

means_fa = []
for i in four_aliens:
    means_fa.append(np.ndarray.mean(np.array(i)))
print("Means of combinations of four aliens: ", means_fa)
```

> iii. (0.3pts) Each of the means above is called a sample mean. Find the mean of all the sample means (denoted by $m_x$) and the standard
deviation of all the sample means (denoted by $\sigma_x$) for both
the *N=2* and *N=4* samples.

```python
import math 

print("N=2, mx: ", np.ndarray.mean(np.array(means_ta)))
print("N=4, mx: ",np.ndarray.mean(np.array(means_fa)))
print("N=2, σx: ", np.ndarray.std(np.array(means_ta)))
print("N=4, σx: ",np.ndarray.std(np.array(means_fa)))

#calculation for question iv
print(np.ndarray.std(x)/math.sqrt(2))
print(np.ndarray.std(x)/math.sqrt(4))
```
