+ *Keep in mind that numpy.linalg.svd() returns the transposed **V<sup>T</sup>** matrix as output.*
+ *You can use **Z** = **Y** $*$ **V**[:,:4] to project the data onto the first four PCs. Don't forget that the $*$ operator does not perform matrix multiplication for numpy arrays!*

```python
## Example code:
#------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.io import loadmat
from numpy import reshape

# Index of the digit to display
i = 0

# Load Matlab data file to python dict structure
mat_data = loadmat('C:\\Users\\rosab\\Documents\\studie 20192020\\Data Mining\\Assignment_2\\Assignment_2\\Data\\zipdata.mat')

# Extract variables of interest
testdata = mat_data['testdata']
traindata = mat_data['traindata']
X = traindata[:,1:]#pixel values
y = traindata[:,0]#digit class

# Remove the digits with the class index 2-9 from the data, so only digits belonging to the class 0 and 1 are analyzed.
# After removal, there are 2199 rows.
ind=0
while ind < len(y):
    if not (y[ind]== 0 or y[ind]==1):
        y=np.delete(y, ind,0)
        X=np.delete(X, ind,0) #0 deletes row, 1 deletes column
    else:
        ind+=1
        
#visualize the first ten digits as an image        
for i in range(10):
    plt.subplot(1,1,1);
    I = reshape(X[i,:],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit ' + str(i+1) + ' as an image');
    plt.show()
      
```

```python
#PCA
# 1. calculate the mean
m = X.mean(axis=0)
# 2. center the data
Y = X - m
# 3. Compute the singular values and vectors
U,sv,Vt = np.linalg.svd(Y) #yields 3 matrices
V = Vt.T
# 4. Calculate the lower dimension representation
m1 = np.dot(Y,V[:,0]) #first principal component
m2 = np.dot(Y,V[:,1]) #second principal component
m3 = np.dot(Y,V[:,2]) #third principal component
m4 = np.dot(Y,V[:,3]) #fourth principal component
#5. project data onto first four PCs
Z = np.dot(Y,V[:,:4])
print(Z)
```

```python
#step4
W = np.dot(Z,V[:,:4].T) + m #is m meant as mean here?

#Use subplotting to show the digits and their reconstructed counterparts in an orderly manner.
plt.figure(1) #to prevent future overlap!
f, axs = plt.subplots(10,2, figsize = (10,10))
axs = axs.ravel() #so that you can assign spots using index 0...19

for i in np.arange(0,20,2): #initial images
    ind = int(i/2)
    I = reshape(X[ind,:],(16,16))
    axs[i].imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    axs[i].title.set_text('Digit ' + str(ind+1))

for i in np.arange (1,20,2): #reconstructed counterparts
    ind = int((i-1)/2)
    I = reshape(W[ind,:],(16,16))
    axs[i].imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    axs[i].title.set_text('Reconstructed digit ' + str(ind+1))
plt.suptitle("Images and reconstructed counterparts", y=1.05, size = 'x-large', weight = 'demibold')  
plt.tight_layout()
plt.show() #don't forget this, otherwise overlap!

# for i in range(10):
#     plt.subplot(1,1,1);
#     I = reshape(W[i,:],(16,16))
#     plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
#     plt.title('Digit ' + str(i+1) + ' as an image');
#     plt.show()
```
