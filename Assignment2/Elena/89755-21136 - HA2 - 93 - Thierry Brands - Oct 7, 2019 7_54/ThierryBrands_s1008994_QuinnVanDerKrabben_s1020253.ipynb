```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sio
import pandas as pd
from scipy import stats

data= sio.loadmat('Data/wine.mat')
Xs = data['X']
X= np.array(Xs)
attr = data['attributeNames']
arrayUnits = ['g/dm^3', 'g/dm^3', 'g/dm^3', 'g/dm^3', 'g/dm^3', 'mg/dm^3', 'mg/dm^3', 'g/cm^3', 'pH', 'g/dm^3', '% vol.' , '0-10']



def plotBox(X, amountSubplots, i, n, outlier): ##Plots a boxplot
    plt.subplot(1, amountSubplots, n)
    new = X[:,i]
    stdNew = stats.zscore(new)
    if (i == 1  or i == 7 or i == 10) and outlier == True:
        plt.title(attr[0][i][0] + ' no outliers')
    else:
        plt.title(attr[0][i][0])
    plt.ylabel(arrayUnits[i])
    plt.boxplot(stdNew)
    plt.tight_layout()

        
def plotHist(X, amountSubplots, i, m, outlier): ##Plots Histogram
    plt.subplot(1, amountSubplots, m)
    new = X[:,i]
    stdNew = stats.zscore(new)
    if (i == 1  or i == 7 or i == 10) and outlier == True:
        plt.title(attr[0][i][0] + ' no outliers')
    else:
        plt.title(attr[0][i][0])
        plt.ylabel('Frequency')
    plt.xlabel(arrayUnits[i])
    plt.hist(stdNew)
    plt.tight_layout()


def plotData(X):
    for i in range(12):
        if i == 1 or i == 7 or i == 10:
            a = findOutliers(X)
            plotBox(X, 2, i, 1, False);
            plotBox(a, 2, i, 2, True);           
            plt.tight_layout()
            plt.show()
            print('These two plots are boxplots of the attribute ' + attr[0][i][0] + 
                  ' from the dataset wine.mat. The left plot is with outliers\n and the right plot is without outliers' '\n\n\n')
            plotHist(X, 2, i, 1, False);
            plotHist(a, 2, i, 2, True);
            plt.tight_layout()
            plt.show()
            print('These two plots are histograms of the attribute ' + attr[0][i][0] + 
                  ' from the dataset wine.mat. The left plot is with outliers\n and the right plot is without outliers' '\n\n\n')
        else:
            plotBox(X, 2, i, 1, False)
            plotHist(X, 2, i, 2, False);
            plt.tight_layout()
            plt.show()
            print('The left plot is a box plot, the right plot is a histogram both of the attribute ' + attr[0][i][0] + 
                  ' from the dataset wine.mat' '\n\n\n')


    
def findOutliers(X):
    volatile_Filtered = X[:, 1] < 20  ## The values of the outliers are 10 times larger than the max expected value
    density_Filtered = X[:, 7] < 10
    alcohol_Filtered = X[:, 10] < 200
    A = []
    
    for i in range(X.shape[0]): ##Function to calculate the amount of outliers and stores the indices of the outliers
        if volatile_Filtered[i] == False or density_Filtered[i] == False or alcohol_Filtered[i] == False:
            A.append(i)
            
    while len(A) > 0: ## The indices are stored in list A. In matrix X the rows of the indices of the outliers are removed.
            b = A.pop()
            X = np.delete(X, b, axis=0) 
    return X
        
plotData(X);
```


Because of the removal of the outliers, it is possible to see more detail in the plots. If the value of the outliers is so big, then the maximum values of the axes of the plot also have to be very big to show the plot including the outliers. Most values of the data are much smaller than the outliers. So if the outliers have a high value, then all other values will be in a small area on the plot because of the scale of the graph has to be so big.

> **Hints:**
+ *The below example code can help you visualize digits as images.*

+ *Keep in mind that numpy.linalg.svd() returns the transposed **V<sup>T</sup>** matrix as output.*
+ *You can use **Z** = **Y** $*$ **V**[:,:4] to project the data onto the first four PCs. Don't forget that the $*$ operator does not perform matrix multiplication for numpy arrays!*
+ *To reconstruct the data from projection you can use the following formula: **W** = **Z**&ast;**V**[:,:4]<sup>T</sup> + ****. *

+ *It is advisable to make a for-loop to generate the 2D scatter plots, this saves a lot of time. It is an important skill to master if you want to easily modify your work later on, for example when correcting mistakes, or when you want to modify each plot in the same manner.* 


```python
## Example code:
#------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.io import loadmat
from numpy import reshape
from mpl_toolkits.mplot3d import Axes3D

# Index of the digit to display
i = 0

# Load Matlab data file to python dict structure
mat_data = loadmat('./Data/zipdata.mat')


# Extract variables of interest
testdata = mat_data['testdata']
traindata = mat_data['traindata']

X = traindata[:,1:]
y = traindata[:,0]

print('Running the code may take some time...')


class0Index=[]
class1Index=[]
class0=[]
class1=[]
classIndex0And1=[]
classIndex01 = y < 2  ##Boolean mask to filter the values 0 and 1

plt.rcParams['font.size'] = 10
B = []
indices01 = []
for i in range (len(y)):    ##Find all class indices above 2
    if classIndex01[i] == False:
        B.append(i)
    

while len(B) > 0:    ##Delete the class indices above 2 from data
            c = B.pop()
            X = np.delete(X, c, axis=0) 
            y = np.delete(y, c)

for i in range (len(y)): ##Find 0 and 1 in data
        if y[i] == 0:
            class0Index.append(i)
        elif y[i] ==1:
            class1Index.append(i)
        
            
            
        
rowVector = []
for i in range (X[0].size):  ##Calculate mean of all columns
    meanOfColumn = np.mean(X[:,i])
    rowVector.append(meanOfColumn)

Y = X - rowVector
U, S, Vt = np.linalg.svd(X)
    
V = Vt.T
Z = np.dot(Y, V[:,:4])##The 4 PC's are in this array of matrices. 


##New lists created here that contain the values of the first 4 PC's for classes 0 and 1
PC1Class0 = []
PC1Class1 = []
PC2Class0 = []
PC2Class1 = []
PC3Class0 = []
PC3Class1 = []
PC4Class0 = []
PC4Class1 = []


for i in range(len(class0Index)):##Makes lists of the PC's that are of class 0
        ci0 = class0Index[i]
        PC1Class0.append(Z[ci0,0])
        PC2Class0.append(Z[ci0,1])
        PC3Class0.append(Z[ci0,2])
        PC4Class0.append(Z[ci0,3])
               
for i in range(len(class1Index)):##Makes lists of the PC's that are of class 1
        ci1 = class1Index[i]
        PC1Class1.append(Z[ci1,0])
        PC2Class1.append(Z[ci1,1])
        PC3Class1.append(Z[ci1,2])
        PC4Class1.append(Z[ci1,3])

##All PC's that belong to one class are combined in 1 list so iteration is possible       
allPCClass0 =[]
allPCClass1 =[]

allPCClass0.append(PC1Class0)
allPCClass0.append(PC2Class0)
allPCClass0.append(PC3Class0)
allPCClass0.append(PC4Class0)

allPCClass1.append(PC1Class1)
allPCClass1.append(PC2Class1)
allPCClass1.append(PC3Class1)
allPCClass1.append(PC4Class1)

W = np.dot(Z, V[:,:4].T) + rowVector  ##Reconstructed images



for i in range (0,10):
# Visualize the i'th digit as an image
    plt.subplot(2,2,1);
    I = reshape(X[i,:],(16,16))
    plt.imshow(I, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image');
    plt.subplot(2,2,2);
    J = reshape(W[i,:],(16,16))
    plt.imshow(J, extent=(0,16,0,16), cmap=cm.gray_r);
    plt.title('Digit as an image reconstructed');
    plt.show()
print('The first 10 digits as an image on the left column and the first 10 digits reconstructed \non the right column\n\n')    


def subplotIndices(j,i): ##Calculate the subplot index of a plot in the 4 by 4 subplot
        index = 0
        if j == 0:
            index = i+1
        elif j==1:
            index = 4 + i + 1
        if j > 1:
            index = 4 * j + i + 1
        return index

PCStrings = ['PC 1', 'PC 2', 'PC 3', 'PC 4']

fig = plt.figure(figsize = (50,50))
for j in range (0,4):
    for i in range (0,4):
        if j == i:
            print()
        else:
            subplotInd = subplotIndices(j,i)
            pl = fig.add_subplot(4,4,subplotInd)
            pl.scatter(allPCClass0[j],allPCClass0[i], label = 'Class 0')
            pl.scatter(allPCClass1[j],allPCClass1[i], label = 'Class 1')
   
            pl.set_xlabel(PCStrings[j], fontsize = 40)
            pl.set_ylabel(PCStrings[i], fontsize = 40) 
            plt.legend(['Class 0','Class 1'], fontsize = 25)
            pl.set_title(PCStrings[j] + ' and ' + PCStrings[i] + ' of mnist digit dataset', fontsize = 40)
            
plt.tight_layout()
plt.show()
print('Scatterplots of the projections of all combinations between PC 1, PC 2, PC 3 and PC 4 from the mnist handwritten digit dataset\n\n')    


##3D plot of first 3 Principle Components
fig3D = plt.figure()
plot3D = fig3D.add_subplot(111, projection = '3d')
plot3D.scatter(allPCClass0[0],allPCClass0[1],allPCClass0[2], label = 'Class0')
plot3D.scatter(allPCClass1[0],allPCClass1[1],allPCClass1[2], label = 'Class1')
plot3D.set_xlabel('      PC 1', fontsize = 20)
plot3D.set_ylabel('      PC 2', fontsize = 20)
plot3D.set_zlabel('      PC 3', fontsize = 20)
plt.legend(['Class 0','Class 1'], fontsize = 12)
plt.tight_layout()
plt.show()
print('3D scatterplot of the projections of the first 3 principle components of the \nmnist handwritten digit dataset\n\n')    


#------------------------------------------------
```


The scatterplots show that class 0 captures more of the variance than class 1. The  scatterplots of class 1 have a higher concentration at the same area.
