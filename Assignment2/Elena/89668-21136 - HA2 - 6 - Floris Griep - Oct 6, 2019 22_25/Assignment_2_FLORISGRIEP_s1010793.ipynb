----
### Calculating the PCA
----

```python
# Calculating the PCA

d = X.mean(axis=0)
e = X.mean(axis=1)

Y = X - d

for o in range (X.shape[1]):
    Y[:,o] = Y[:,o] - e[o]
    
u, s, v = np.linalg.svd(Y)
```

----
### Calculating Z
----

```python
Z = np.dot(X, v.T[:,:4])

print(Z.shape)
print(X.shape)

plt.subplot(1,1,1);
I = reshape(Z[1,:],(2,2))
plt.imshow(I, extent=(0,2,0,2), cmap=cm.gray_r);
plt.title('Image');
plt.show()
```

----
### Calculating W
----

```python
import pandas as pd
import seaborn as sb

# The reconstruction of W

W = np.dot(Z, (v[:,:4]).T)+d


for g in range (X.shape[1]):
    W[:,g] = W[:,g] + e[g]
```

```python
alienLimbs = np.array([2,3,6,8,11,18])

print(alienLimbs.mean())
print(alienLimbs.std())
```

> ii. (0.3pts) List all possible samples of two aliens without replacement, and find each mean. Do the same with samples of four aliens.
+ *You can use the method itertools.combinations(v,n) to find all possible samples of a vector v taking n elements at a time.*

```python
import  itertools as it

for combo in it.combinations(alienLimbs,2):
    print(combo,(sum(combo))/len(combo))

for combo in it.combinations(alienLimbs,4):
    print(combo,(sum(combo))/len(combo))    
```

> iii. (0.3pts) Each of the means above is called a sample mean. Find the mean of all the sample means (denoted by $m_x$) and the standard
deviation of all the sample means (denoted by $\sigma_x$) for both
the *N=2* and *N=4* samples.

```python
two = []
i = 0
for combo in it.combinations(alienLimbs,2):
    two.append((sum(combo))/len(combo))
    i += 1

four = []
i = 0
for combo in it.combinations(alienLimbs,4):
    four.append((sum(combo))/len(combo))    
    i += 1
```

```python
two = np.array(two)
print(two.mean())
print(two.std())
```

```python
four = np.array(four)
print(four.mean())
print(four.std())
```
