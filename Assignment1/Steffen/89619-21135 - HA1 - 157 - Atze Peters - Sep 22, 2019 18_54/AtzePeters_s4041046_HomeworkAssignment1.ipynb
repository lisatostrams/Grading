
1.2.2 Principal Component Analysis, more commonly known as PCA, can be used to visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

--------

_write your answer to 1.2.2 here_

Principal Component Analysis (PCA) is a method to find a set of attributes that better captures the variability of the data compared to the original set of attributes. The amount of variability captured by successive attributes decreases as as much of the variability is tried to be captured in each successive attribute, while being orthogonal to the previous attributes.
As PCA tends to identify patterns that are present in the data, it can be used as a pattern-finding technique. On top of that it can also be used as a technique for dimensionality reduction as most of the data can often be captured in only a few attributes. Lastly, it also reduces the noise of the data as the noise is often weaker than the patterns within the data.
Some of the downsides are the lessened interpretability of the pricipal components as compared to the original attributes. Besides this, the number of principle components wanted, should be chosen with care as information may end up lost when the wrong number is chosen.

-------

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

ANTWOORD: The step of mean centering the data is a normalization step to centre the data around the origin.

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute. Then you can subtract **μ** from **X** (**Y** = **X** - **μ**, where **μ** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
mu = X.mean(0)
Y = X - mu
U,sv,Vt = np.linalg.svd(Y)
V = np.transpose(Vt)
Z = np.dot(Y,V)

plt.scatter(Z[:,0],Z[:,1])
plt.xlabel('Projection on 1st PC')
plt.ylabel('Projection on 2nd PC')
plt.title("Scatter plot of projections on 1st and on 2nd PC")
print('Figure 5: A scatter plot showing the projections of the 1st(x-axis) and 2nd(y-axis) principle components')
```

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

ANTWOORD: The most important difference between EVD and SVD is that EVD uses eigenvalues. Eigenvalues are a property of square matrices, and square matrices alone. This means that EVD cannot be used on arbitrary matrices that aren't square matrices. SVD, on the other hand, can be use on any arbitrary matrix.

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values. 
> e. (1 point) Calculate the row vector $\rho$ containing all values of $\rho$ for all PCs **m**. Create a bar plot with the variance explained on the Y axis and the number of the PC on the X axis. 

```python
quot = 1/np.sum(sv**2)
rho = quot*sv*sv
index = np.arange(start=1,stop=9)
plt.bar(index,rho)
plt.xlabel('nth PC')
plt.ylabel('explained variance')
plt.title('Explained variance by the nth PC')
print("Figure 6: A bar chart depicting the variance explained by each of the PC's. As you can see, the explained variance decreases for each successive PC.")
```

The columns of **V** indicate the exact projection of the data onto the PCs. In a way, a PC is nothing other than a linear combination of the original attributes.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

```python
from Toolbox.similarity import similarity
import numpy as numpy

#the following will be the parameters for calculation
x = np.array([0,7,8])
y = np.array([5,8,2])
alpha = 5
beta = 5

print("similarity measures for cosine similarity")
print("cos(x,y)")
print(similarity(x,y,'cos'))
print("cos(ax,y)")
print(similarity(alpha*x,y,'cos'))
print("cos(b+x,y)")
print(similarity(beta+x,y,'cos'))
print("the fact that the first two are the same is to be expected as the cosine similarity calculates the angle between the vectors. When one of the vectors is multiplied with a constant, only the length changes, leaving the angle constant. When a constant is added to one of the vectors, the direction of that vector changes, changing the angle between the two vectors in the process")
print("-")
print("similarity measures for extended Jaccard similarity")
print("ext(x,y)")
print(similarity(x,y,'ext'))
print("ext(ax,y)")
print(similarity(alpha*x,y,'ext'))
print("ext(b+x,y)")
print(similarity(beta+x,y,'ext'))
print("For this similarity measure, both addition and multiplication by a scalar changes the similarity. This is to be expected as the measure looks at the attributes of the vector where both are non-zero. When one of the vectors is multiplied by a scalar >1, the two become less similar because of increase in lenght. The length of the vectors is in the denominator of the formula for calculating the extended Jaccard. The measure also becomes different for addition with a scalar as the angle between the two vectors will change, present in both the numerator and denominator.")
print("-")
print("similarity measures for correlation similarity")
print("cor(x,y)")
print(similarity(x,y,'cor'))
print("cor(ax,y)")
print(similarity(alpha*x,y,'cor'))
print("cor(b+x,y)")
print(similarity(beta+x,y,'cor'))
print("The correlation similarity measure, measures the similarity by first normalizing the vectors. this ensures an equal similarity for two vectors even if scalar addition or multiplication is used on one the two vectors")
```
