
1.2.2 Principal Component Analysis, more commonly known as PCA, can be used to visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

--------

PCA, or Principal Component Analysis, is a linear technique that is used to rotate the axes in a new position, where these new rotated axes are ordered by the variance (the first principal axis has the highest variance of the data, the last one has the lowest variance) and they are uncorraleted from each others.
This means that the first principal components contain the major part of the information, while the last ones only a little.
So if we remove the last few components we will lose only small information; we can calculate that amount of information lost adding all the corresponding eigenvalues of the eigenvectors of those principal components we want to remove (all eigenvalues' components will sum up at 1, so a sum of 0.1 is like 10% of variance and information lost).

Using PCA and then removing last components is the best way to reduce the dimensionality losing the lowest amount of information. This is important when we want to reduce the dataset's size and the computing time of an algorithm, or to remove not relevant feautures or noise (when data are taken from sensors).
PCA is also a good way to visualize in a better way the data, plotting the first principal components.

The principal limitation of PCA is that is time consuming with large datasets (for the matrix operations) and it is unfeasible if you have a lot of feautures (the covariance matrix will be too large).
Another limitation is that is a linear transformation with some assumptions (orthogonal axes, ecc.) so it may perform bad compared to a non-linear algorithm (this depends on which kind of data do you have).


-------

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

------

Mean centering the data is applied to center the data in the origin of the axes. This is done because PCA is a rotation (and not a traslation) of the axes, so if the data are not centered it may not be able to follow the points in the highest variance's direction with the first principal component.
Another advantage we gain is that centering the data lead to simpler calculus and with lower numbers.
PCA also presupposes that data are centered; if you don't subtract the mean you are just doing an SVD.

-------------

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute. Then you can subtract **** from **X** (**Y** = **X** - ****, where **** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
mean = np.mean(X, axis=0)

Y = X - mean

U, S, VT = np.linalg.svd(Y)


V = VT.T
Z = np.dot(Y, V)

plt.scatter(Z[:, 0], Z[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('plot of the first two Principal Components of PCA')
plt.show()
```

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

---

The first difference between EVD and SVD is on the requirements: SVD don't need any requirements and can be used with all the rectangular matrixes, while EVD needs a diagonalizable matrix. So we can say that SVD is more powerful than EVD.
Another difference is that with EVD we will have only one basis, while SVD gives you one left and one right.

If we want to use EVD on a non-square matrix first we have to compute the covariance matrix and then compute the EVD on it. These operations will cost more compared to an SVD and generally SVD is numerically more stable rather than EVD because it doesn't need to compute the covariance matrix. There are also some seldom cases where computing the covariance matrix lead to numerical problems.



So generally they can both be used for the PCA, but for a basic PCA the SVD would be the better option.

---

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values. 
> e. (1 point) Calculate the row vector $\rho$ containing all values of $\rho$ for all PCs **m**. Create a bar plot with the variance explained on the Y axis and the number of the PC on the X axis. 

```python
S_squared = np.power(S, 2)
sum = np.sum(S_squared)

#print(S, S_squared, sum)

rho = S_squared / sum

print(rho)

plt.bar(np.arange(8), rho)
plt.title('Bar plot of the 8 variances')
plt.show()
```

```python
sumvar = np.sum(rho[0:3])
print(sumvar)
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

> Even though a similarity measure is theoretically invariant e.g.\ to scaling, it might not be exactly invariant numerically.

> Do not forget to also provide an explanation in addition to the calculations!


```python
from Toolbox import similarity as sim

x = np.random.rand(10)
y = np.random.rand(10)
#print(x,"\n", y)

a = np.random.rand()
b = np.random.rand()

#print(x, "\n", a, "\n", x * a)
print("let's see what happen if we scale the data\n")
cos_axy = sim.similarity(a * x, y, 'cos')
cos_xy = sim.similarity(x, y, 'cos')
print("the difference between cos(ax, y) and cos(x, y) is: ", cos_axy - cos_xy)

ext_axy = sim.similarity(a * x, y, 'ext')
ext_xy = sim.similarity(x, y, 'ext')
print("the difference between ext(ax, y) and ext(x, y) is: ", ext_axy - ext_xy)

cor_axy = sim.similarity(a * x, y, 'cor')
cor_xy = sim.similarity(x, y, 'cor')
print("the difference between cor(ax, y) and cor(x, y) is: ", cor_axy - cor_xy)

print("\n\nlet's see what happen if we translate the data\n")
cos_bxy = sim.similarity(b + x, y, 'cos')
print("the difference between cos(b + x, y) and cos(x, y) is: ", cos_bxy - cos_xy)

ext_bxy = sim.similarity(b + x, y, 'ext')
print("the difference between ext(b + x, y) and ext(x, y) is: ", ext_bxy - ext_xy)

cor_bxy = sim.similarity(b + x, y, 'cor')
print("the difference between cor(b + x, y) and cor(x, y) is: ", cor_bxy - cor_xy)
```

---

Cosine is not affected from the scaling (as we can see the difference is really close to zero) but is a bit affected from the translation.
This is because during scaling the alpha can be simplified on the numerator and denominator.

Extended Jaccard is affected both from scaling and translation, with a difference from the normal similarity (without a or b) that can be large.
That happens because if you try to insert a scaling or translation inside the formula you will find, after the calculations,  that it different from the original one.

correlation is the only one that is not impacted by translation neither scaling.
This is because both covariance and standard deviation are invariant to scaling and translation.
