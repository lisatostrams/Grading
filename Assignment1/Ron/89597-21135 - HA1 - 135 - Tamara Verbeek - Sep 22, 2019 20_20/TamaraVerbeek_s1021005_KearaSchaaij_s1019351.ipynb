
1.2.2 Principal Component Analysis, more commonly known as PCA, can be used to visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

--------
Principle component analysis finds the principlar components of the data, it can be used to reduce a large set of variables to a small set that still contains the most important information. It does this by converting a set of correlated variables into principal components, these components have the biggest variance. The principal components are a set of linearly uncorrelated values. 
You start with finding the first principal component, this is the direction with the largest variability in the data as possible.
After this, you have to find the second principal component. This is the direction orthogonal to the principal component, this one has the second largest variation. And so on.
Then these principal components are the eigenvectors of the covariance matrix. They are ordered by their corresponding eigenvalues. The most variation is the biggest eigenvalue.
After this, we have to project the data points onto the first few principal components.

We use principal component analysis when we have data with too many dimensions which cause noise and difficulties.

limitations of PCA are:
- If the principle components aren't a linear combination of the original features, PCA will not give you a reliable result
- It is not always the case that the dimension with the highest variance is the most important. 
- The principal components are expected to be orthogonal, otherwise the PCA doesn't work.
-------

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

------
At first we must mean center the data, before we apply PCA. This is because covariance matrix is sensitive to standardization of variables. If the variables are not standardized before applying principle component analysis, we might get misleading directions.  

------


In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute. Then you can subtract **** from **X** (**Y** = **X** - ****, where **** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

---

Differences between SVD and EVD:
1. You can do EVD only with square matrices. Sometimes EVD doesn't even exist. However, SVD always exists, also for non-square matrices. 
2. In SVD, the numbers can only be non-negative while the numbers in EVD can also be negative.
3. In EVD the vectors don't have to be orthonormal, while in SVD the vectors in the matrices are orthonormal. 
4. In EVD, the non-diagonal matrices are inverses of each other, while in SVD this is not (necessarily) the case. 

Similarities between SVD and EVD:
1. EVD and SVD are factorizations of a matrix.

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

```python
import math
from numpy import array
import numpy as np
from scipy import stats

X = array([1, 2, 3])
Y = array([4, 5, 6])
M = np.shape((Y) -1)
a = 3
b = 10


#Cosine(x,y)
cos1 = (X*np.transpose(Y))/np.transpose(np.sqrt(sum(np.power(np.transpose(X),2))) * np.sqrt(sum(np.power(np.transpose(Y),2))))
print("Cosine(x,y)")
print(cos1)

#Cosine(,)
cos2 = ((X*a)*np.transpose(Y))/np.transpose(np.sqrt(sum(np.power(np.transpose(X*a),2))) * np.sqrt(sum(np.power(np.transpose(Y),2))))
print("Cosine(ax,b):")
print(cos2)


XYt = np.transpose(X*Y)

#ExtendedJaccard(x,y)
jac1 = XYt / np.transpose(np.log( np.exp(sum(np.power(np.transpose(X),2))) * np.exp(sum(np.power(np.transpose(Y),2))) ) - XYt)
print("ExtendedJaccard(x,y)")
print(jac1)

#ExtendedJaccard(ax,y) 
aXYt = np.transpose((a*X)*Y)
jac2 = aXYt / np.transpose(np.log( np.exp(sum(np.power(np.transpose(a*X),2))) * np.exp(sum(np.power(np.transpose(Y),2))) ) - aXYt)
print("ExtendedJaccard(ax,y)")
print(jac2)


Q = np.array([[7, 8],
              [11, 12]])
H = np.array([[4, 5],
              [1, 2]])

#Correlation(q,h)
Q_ = stats.zscore(Q,axis=1,ddof=1)
H_ = stats.zscore(H,axis=1,ddof=1)

cor1 = (Q_*np.transpose(H_))/ (M)
print("Correlation(q,h)")
print(cor1)

#Correlation(aq,h)
aQ_ = stats.zscore((a*Q),axis=1,ddof=1)
H_ = stats.zscore(H,axis=1,ddof=1)
cor2 = (aQ_*np.transpose(H_))/(M)
print("Correlation(aq,h)")
print(cor2)

#Cosine((b + x),y)
cos3 = ((b+X)*np.transpose(Y))/np.transpose(np.sqrt(sum(np.power(np.transpose(b+X),2))) * np.sqrt(sum(np.power(np.transpose(Y),2))))
print("Cosine((b+x),y)")
print(cos3)
print("Cosine(x,y)")
print(cos1)

#ExtendedJaccard((b+x),y)
bXYt = np.transpose((b+X)*Y)
jac3 = bXYt / np.transpose(np.log( np.exp(sum(np.power(np.transpose(X+b),2))) * np.exp(sum(np.power(np.transpose(Y),2))) ) - bXYt)
print("ExtendedJaccard((b+x),y)")
print(jac3)
print("ExtendedJaccard(x,y)")
print(jac1)


#Correlation(b+q,h)
bQ_ = stats.zscore((Q+b),axis=1,ddof=1)
H_ = stats.zscore(H,axis=1,ddof=1)

cor3 = (bQ_*np.transpose(H_))/ (M)

print("Correlation(q,h)")
print(cor1)

print("Correlation(b+q,h)")
print(cor3)
```

Cosine(x,y) and Cosine(ax,y)

We have the formula for calculating the similarity of cosine. This formula is for Cosine(x,y), when you want to calculate Cosine(ax,Y) you have to replace every X in the formula by (X*a). In this way you are able to calculate Cosine(ax,y). In the original formula there was a "x.t" that this programm did not recognise. That is why we used np.transpose(x) instead.

For Extended Jaccard(x,y) and Extended Jaccard(ax,y) we did the same thing as above.

For Correlation(q,h) and Correlation(aq,h) we did the same thing as above, however there were a few more parts from the original formula that we were not able to use in this programm. We had to use the "zscore" instead we use "stats.zscore(..)". There was a variable M that we used in the formula and we had to substract 1 from it, which was also not possible. We solved this by "M = np.shape((Y) -1)".

For Cosine(x,y) and Cosine(b+x,y) we again used the original formula of Cosine(x,y). Now we have replaced X by (b+X). The problem of the "x.t" occured again so we solved this at the same way as before. Just change it to np.transpose(X).

For Extended Jaccard(b+x,y) and Extended Jaccard(x,y) we did the same thing as above.

For Correlation(q,h) and Correlation(aq,h) we did the same thing as above, again there were a few more parts from the original formula that we were not able to use in this programm. We had to use the "zscore" instead we use "stats.zscore(..)". There was a variable M that we used in the formula and we had to substract 1 from it, which was also not possible. We solved this by "M = np.shape((Y) -1)".
