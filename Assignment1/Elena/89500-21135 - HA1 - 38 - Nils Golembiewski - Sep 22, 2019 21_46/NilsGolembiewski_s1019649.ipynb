
1.2.2 Principal Component Analysis, more commonly known as PCA, can be cleverly used to better visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

PCA is a method which uses linear algebra to split the data into 'principal components' which are linear combinations of the data. These principal components are calculated in such a way that the first principle component has the largest variation possible. In order to calculate them, the data must first be mean centered. Then the data is rotated in such a way that the first axis has the largest variation, followed by the second, and so on. These axes are the principle components. The rotation is done using singular value decomposition. The useful part about the data is that the first principle component tells us the most about the data(, followed by the second, third, ...). So in order to reduce dimensions, a set number of principle components can be used. For instance, when the original dimension is 9, and the first 3 principal components are taken, it reduces computation time a lot when analyzing the data while still retaining a large amount of information. So it is useful in scenarios where highly dimensional data is collected, possibly with a lot of noise. One drawback of principle component analysis is that it fails in detecting non-linear correlated data (see https://arxiv.org/abs/1404.1100 figure 6). Another drawback is that the data first has to be mean centered. This might introduce negative values for data which originally did not have negative values.

To apply PCA we must first mean center the data. mean centering means that the mean value for an attribute is subtracted from that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

A principle component goes through the origin. So, when the data is not mean-centered, a wrong principal component might be chosen, because there might be a line possible through the data which has more variation, but it does not go through the origin. If the data is mean-centered, such a line will always go through the origin, and thus will be the principal component. This drawing illustrates it: https://imgur.com/a/WM27enO

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X**, must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute and then subtracting it from **X** (**Y** = **X** - **1μ**, where **1** is a column vector of size **N** x 1, with N indicating the number of observations in the original data).

Then, the Singular Value Decomposition, or SVD, of **Y**, the mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the numpy.linalg.svd() function.

Using SVD on **Y** yields a series of vectors which can be used to project the data onto specific Principal Components (PCs). The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,1]** would yield the projection of the data onto the first PC.

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
# Calculate mean
mean_nnn = nanonose_values.mean(0)
# Transform nanonose values to center
centered_nnn = nanonose_values - mean_nnn
# Calculate the SVD
u, singular_values, vt = np.linalg.svd(centered_nnn)
# Calculate original v
v = vt.transpose()

plt.title("Projection of the nanonose values onto the first two principal components")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
#Plot first pc against second pc
plt.scatter(centered_nnn.dot(v[:,0]), centered_nnn.dot(v[:,1]))
plt.show()



```

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

The formula for Eigen Value decomposition is A = UΛU<sup>T</sup> for Singular Value Decomposition this is  A = SΣV<sup>T</sup>. In Eigenvalue decomposition, U is obviously the same size as U<sup>T</sup>, but in SVD, S does not have to be the same size as U<sup>T</sup>. Both SVD and EVD have diagonal matrix containing values from large to small, where in EVD they are the eigen values, and in SVD they are the singular values. Both can be used for PCA, but one large drawback of EVD is that it can only be used on square matrices, while SVD can be applied to any matrix. When working with large datasets, it is very unlikely that a data matrix is square.

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values. 
> e. (1 point) Calculate the row vector $\rho$ containing all values of $\rho$ for all PCs **m**. Create a bar plot with the variance explained on the Y axis and the number of the PC on the X axis. 

```python
# Calculate sum of singular values
sum_squared_singular_values = sum([sv**2 for sv in singular_values])
# Calculate the rho vector
rho_v = np.array([(sv**2)/sum_squared_singular_values for sv in singular_values])

plt.title("Explained variance for each principal component")
plt.xlabel("Number of principal component")
plt.ylabel("Explained variance")
plt.bar(range(len(rho_v)+1)[1:], rho_v)
plt.show()
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 gray scale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\end{eqnarray*}
where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary
data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary,
we will transform the data to be binary when calculating these two measures of
similarity by setting
$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$

> (0.75 points) We will investigate how scaling and translation impact the following three
similarity measures: Cosine, ExtendedJaccard, and Correlation. Let $\alpha$ and $\beta$ be two constants. Using Python, calculate the following similarity measures, and check if the statements below are correct.

\begin{eqnarray*}
\textrm{Cosine}(\alpha{x},{y}) & = & \textrm{Cosine}({x},{y}) \\
\textrm{ExtendedJaccard}(\alpha{x},{y}) & = & \textrm{ExtendedJaccard}({x},{y}) \\
\textrm{Correlation}(\alpha{x},{y}) & = & \textrm{Correlation}({x},{y}) \\
\textrm{Cosine}(\beta + {x},{y}) & = & \textrm{Cosine}({x},{y}) \\
\textrm{ExtendedJaccard}(\beta + {x},{y}) & = & \textrm{ExtendedJaccard}({x},{y}) \\
\textrm{Correlation}(\beta + {x},{y}) & = & \textrm{Correlation}({x},{y})
\end{eqnarray*}

> Type help similarity, or study similarity.py, to learn about the function that is used to compute the similarity measures. Do not forget to also import similarity.py!

> Even though a similarity measure is theoretically invariant e.g.\ to scaling, it might not be exactly invariant numerically.

> Do not forget to also provide an explanation in addition to the calculations!


```python
from Toolbox.similarity import similarity

# Generate range of constants
alpha = np.arange(1,20,.2)
beta = np.arange(0,100,1)

# Generate matrices with ones and zeros
x_matrix = np.random.randint(2,size=(10, 10))
y_matrix = np.random.randint(2,size=(10, 10))

methods = ["Cosine", "ExtendedJaccard", "Correlation"]

def mult_fn(a, b):
    return a * b

def add_fn(a,b):
    return a + b

def diff(constants, method, operator_fn):
    difference_sums = []
    # Generate normal similarity matrix
    sim_matrix = similarity(x_matrix, y_matrix, method)
    for constant in constants:
        difference=0
        # Generate a similarity matrix with the constant
        const_sim_matrix = similarity(operator_fn(constant, x_matrix), y_matrix, method)
        #compare each value and add the absolute difference to the sum of total differences
        for x, y in np.ndindex(sim_matrix.shape):
            difference += np.absolute(sim_matrix[x,y]- const_sim_matrix[x,y])
        difference_sums.append(difference)
    return difference_sums

for method in methods:
    plt.title(f"Sum of differences between {method}(alpha * x, y) and {method}(x, y)\n")
    plt.xlabel("alpha")
    plt.ylabel("Sum of differences")
    plt.plot(alpha, diff(alpha, method, mult_fn),'.')
    plt.show()
    
    plt.title(f"Sum of differences between {method}(beta + x, y) and {method}(x, y)\n")
    plt.xlabel("beta")
    plt.ylabel("Sum of differences")
    plt.plot(beta, diff(beta, method, add_fn),'.')
    plt.show()
    
    
    
```

That sum is what is visible in the plots. In some plots, the value of this sum is always a very small number, this is just due to the rounding of floating point numbers. In other plots there is a difference visible between the original and the scaled version.<br>
