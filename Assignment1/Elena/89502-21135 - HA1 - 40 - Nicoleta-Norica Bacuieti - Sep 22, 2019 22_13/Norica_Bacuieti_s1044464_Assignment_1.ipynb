
1.2.2 Principal Component Analysis, more commonly known as PCA, can be used to visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

-------------------------------------------------------------------------------------------------------------------------------


PCA is defined as the eigendecomposition of the covariance matrix X.transpose() * X where X is our data. It allows us to reduce the high-dimensional data into something that can be explained into fewer dimensions so that we can better visualize it. The eigendecomposition gives us some eigenvectors and eigenvalues and we use those to describe our data in another system of orthogonal axes. Each eigenvector is a principal component ordered in decending order form left to right according to its corresponding eigenvalue. In this way, the first PC is going to be in the direction that explains/captures the most variance of the data.The second PC is orthogonal to the first, and captures as much of the remaining variability as possible, and so on.  In addition, each eigenvalue lambda_i, tells us how much variance is in the data along the i'th dimension. PCA can be viewed as a rotation (and translation) of the original coordinate axes to a new set of axes that are aligned with the variability in the data. The total variability of the data is preserved, but the new attributes are uncorrelated. Also, PCA does not change the data. It removes correlated features and reduces noise. 

For example: If we want to visualize our data X in two dimensions, then we keep just the first two PC, and then multiply our data matrix X with them. In this way we will get a two-dimensional representation of the SAME data. 

However, there are some drawbacks. 
The data must be standardized before applying PCA, otherwise PCA won't be able to find the optimal principal components. This has to be done so that the eigenvectors origin will be in the center of the data, and not computed starting from the origin of the initial axes of coodinates. It is done so that the origin of the XY plane is moved in the center of our data. By this we also bring the values on the same scale of measurements; that means that we give each attribute a fair chance to influence the results of computing the principal components. In this way, each variation of the attributes will be captured and given importance according to their relative variation.   

In addition to that, if we do not choose the right number of PC's then information will be lost (maybe more than we wanted). 

A limitation of PCA in classifying data: 
Let's say that we have some red and blue dots (in 2 dimensions) that are (visually )separated along the direction with the highest variance. We apply PCA, project the data on one dimension and choose the first PC to classify the data. And it does a god job! 
But, if our data was separated along the direction of the lowest variance, if we would choose the first PC (as usual) to classify the data (after we projected it on one dimension) it will not classify the data effectively. The best choice would be to classify the data using the second PC. Here we had labels and could see and make a choice but if you do not have labels, just assuming that the first PC will be the best choice does not guarantee you that it will give the best results.

Why mean center before applying PCA?
As I said earlier, we do that to bring the center of the axes of coordinates in the center of the data so that the variance will be considered starting from that new refference. In this way, the principal components reflect(and capture) the true variance of the data. Just mean centering the data does not change the scale of measurement. To change the scale as well as to move the origin of the XY axes, you will have to standardize your data.  


-------------------------------------------------------------------------------------------------------------------------------


In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute. Then you can subtract **** from **X** (**Y** = **X** - ****, where **** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

The Cosine Coefficient computes similarity based on the proportion of the present attributes; that means that two vectors are similar if they contain the same attributes in the same proportion ( it is concerned with the degree to which the two vectors point in the same direction). The magnitude of the vectors is not taken into account. When you multiply a vector with a scalar, the direction of the vector does not change, its magnitude just gets bigger. That is why COS(alpha*x,y)=COS(x,y). But when you add a scalar to each attribute, it it like you added another vector to it. This causes the direction of the initial vector to change, and thus give different results. Therefore, COS(beta+x,y) != COS(x,y). 

The Extended Jaccard Coefficient computes similarity based on the presence of the attributes ( 00 matches are not considered) and magnitude of the vectors. Therefore, the multiplication and addition with a scalar influences the results. 

The Correlation Coefficient standardizes the vectors. The ideea is that you take the mean from the values and divide them by the standard deviation. The difference (x - mean_x) and the standard deviation are invariant. (I also read that the correlation coefficient is invariant to change of origin and scale of the vectors, and standardization will not alter the correlation.) Therefore, the results do not change.
