
1.2.2 Principal Component Analysis, more commonly known as PCA, can be used to visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

### Answer to question 1.2.2a

--------
The Principal Component Analysis, also known as PCA, is one of the statistical techniques which allow to reduce the dimensionality of a matrix (i.e. variables, if we are talking about Data frames).
Typically, the number of calculations we need to perform on the data increase on a quadratic scale with every dimension (variables, columns), making it computationally demanding.
We can prove many reasons about the utility of this methods. Mainly, we need to reduce dimensionality in order to:
- decrease computational time and memory of data mining algorithms,  
- reduce data noise
- allow us to easily visualize data

It has to be noted that PCA does not perform feature elimination, but it creates fewer new features, also called Principal Components. 
This process is carried on through a linear transformation of the variables within the data. The transformation yields to a new Cartesian coordinate system, in which the Principal Components, uncorrelated (i.e. orthogonal) to each other, are sorted in variance decreasing order.
The complexity reduction occurs by considering only the PCs that explain greater percentage of variance.
The whole method is designed to well control the trade-off between information loss and problem simplification.


Let the matrix of the original data **X** be of np size, where n is the number of records and p is the number of variables. Let us assume that this matrix is centered, so that the column mean is subtracted from each element.

Let now the covariance matrix **E** be pp and such that **E = $\frac{X^TX}{(n1)}$**. 
It is a squared and symmetric, thus it can be diagonalized:
**E = VLV<sup>T</sup>**,

where:
- **V** is a matrix of eigenvectors (each column is an eigenvector).
- **L** is a diagonal matrix with eigenvalues $\lambda_i$ on the diagonal in decreasing order. 

The eigenvectors are called principal axes of the data. Projections of the data on the principal axes are the already mentioned Principal Components, which can be seen as new, transformed variables. 
The j-th PC is given by j-th column of **XV**. 
The coordinates of the i-th data point in the new Principal Component space are given by the i-th row of **XV**.

Now, if we perform the singular value decomposition of **X**, we obtain a decomposition
**X = USV<sup>T</sup>**,
where:
- **U** and **V** are unitary matrices.
- **S** is the diagonal matrix of singular values $s_i$. 

Keeping that in mind, we can see that:
$$ E= \frac{VSU^T USV^T}{n1} = V \frac{S^2}{n1} V^T $$

The right singular vectors V are the principal axes and the singular values are related to covariance matrix's eigenvalues by $ \lambda_i = \frac{{s_i}^{2}}{n  1} $. 
Then, Principal Components are also given by

**XV = USV<sup>T</sup>V = US** .

In summary, we can say that:

- If **X=USV<sup>T</sup>**, then columns of V are principal axes.

- Columns of **US** are the Principal Components.

- Singular values are related to the eigenvalues of the covariance matrix via $ \lambda_i = \frac{{s_i}^{2}}{n  1} $. Eigenvalues $\lambda_i$ show variances of the respective PCs.


- The process above is correct only if **X** is centered. Only then the covariance matrix is equal to X<sup>T</sup>X/(n1), and obviously if **X** has records in rows and variables in columns.

- It's possible to perform PCA on a correlation matrix (instead of a covariance matrix), standardizing columns of **X**. This process could be useful when data don't have the same scale.

- To reduce the dimensionality of the data from p variables to k < p PCs, we need to select the k first columns of **U**, and kk top-left part of **S**. Their product $ U_k S_k $ is the required nk matrix containing first k PCs.



We could say that PCA has not really disadvantages, but some limitations due to the starting assumpions.
As wrote above, it yields to an information loss, which is compensated by simplification.
The creation of new variables, less then the original ones, leads to a more difficult interpretation of the data. 
Also, PCA assumes that the Principle Components are a linear combination of the original features; if this is not true, PCA will not give us sensible results.
Finally, PCA uses variance as the measure of how important a particular column or dimension is. This assumption might not be useful for some problems.

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

### Answer to question 1.2.2b

------
The entire the process of PCA, explained above, is correct only if the matrix of original data **X** is centered.
In particular, mean centering is necessary to ensure that the first PC describes the direction of maximum variance. If mean subtraction is not performed, the first PC might instead correspond approximately to the mean of the data.

Actually, PCA is equivalent to SVD if we don't center the data; the gist is the simplicity about covariance matrix calculation.
As a matter of fact, if the data are centered, we can compute the covariance matrix as:

$$E = \frac{X^TX}{n1}$$

That being said, PCA maximizes (by the 1st PC) sum-of-squared deviations from the origin. Only if the data were preliminary centered (centering itself isn't a part of PCA) it turns to be maximizing variance.

-----------------------------------------

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute. Then you can subtract **** from **X** (**Y** = **X** - ****, where **** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

### Answer to question 1.2.2d

---
**EVD** is a process to factorize diagonalizable and squared matrices.

Let A be a square n  n matrix with n linearly independent eigenvectors $q_i$ (where i = 1, ..., n). 
Then A can be factorized as:

$$ A = Q \Lambda Q^{-1} $$

where Q is the square nn matrix whose ith column is the eigenvector $q_i$ of A, and  is the diagonal matrix whose diagonal elements are the corresponding eigenvalues, $\Lambda_{ii} = \lambda_i$. 
We need to emphasize that only diagonalizable matrices can be factorized in this way.

The **SVD** is another method to factorize matrices.
It could be defined as a generalization of EVD seen above, to any kind of matrix, so that:

$$ M = U \Sigma V^T $$

as described above.


The differences are:

- While SVD uses two different basis, which are called *left* and *right singular vectors*, the EVD uses only the eigenvector as basis. 
- The basis of the SVD is orthonormal, while the ones of the EVD not necessarly orthogonal.
- PCA can be performed by EVD of a data covariance matrix or by SVD of a data matrix, after the mean-centering step.



Finally, while we can state that every matrix can be factorized by the SVD method, on the other hand, not even every square matrix has an eigendecomposition.

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

```python
from pylab import *
from scipy.io import loadmat
import pathlib
from Toolbox.similarity import similarity

def simfaces(i=1, similarity_measure = 'smc'):
    # i is the Image to use as query

    # Similarity: 'SMC', 'Jaccard', 'ExtendedJaccard', 'Cosine', 'Correlation' 


	# Load the CBCL face database
	# Load Matlab data file to python dict structure
    #str(pathlib.Path.cwd().joinpath("Data","wildfaces_grayscale.mat"))
	X = loadmat("Data/wildfaces_grayscale.mat")['X']
	N, M = shape(X)


	# Search the face database for similar faces
	# Index of all other images than i
	noti = list(range(0,i)) + list(range(i+1,N))
	# Compute similarity between image i and all others
	sim = similarity(X[i,:], X[noti,:], similarity_measure)
	sim = sim.tolist()[0]
	# Tuples of sorted similarities and their indices
	sim_to_index = sorted(zip(sim,noti))


	# Visualize query image and 5 most/least similar images
	figure(figsize=(12,8))
	subplot(3,1,1)
	imshow(np.reshape(X[i],(40,40)).T, cmap=cm.gray)
	xticks([]); yticks([])
	title('Query image')
	ylabel('image #{0}'.format(i))


	for ms in range(5):

		# 5 most similar images found
		subplot(3,5,6+ms)
		im_id = sim_to_index[-ms-1][1]
		im_sim = sim_to_index[-ms-1][0]
		imshow(np.reshape(X[im_id],(40,40)).T, cmap=cm.gray)
		xlabel('sim={0:.3f}'.format(im_sim))
		ylabel('image #{0}'.format(im_id))
		xticks([]); yticks([])
		if ms==2: title('Most similar images')

		# 5 least similar images found
		subplot(3,5,11+ms)
		im_id = sim_to_index[ms][1]
		im_sim = sim_to_index[ms][0]
		imshow(np.reshape(X[im_id],(40,40)).T, cmap=cm.gray)
		xlabel('sim={0:.3f}'.format(im_sim))
		ylabel('image #{0}'.format(im_id))
		xticks([]); yticks([])
		if ms==2: title('Least similar images')
		
	show()
```

As we can see from the above calculations Cosine measure is invariant for scaling, while the translation of the vector x yields to different numerical results for the same measure.
In fact for the scaling:
\begin{eqnarray*}
\textrm{Cosine}({\alpha x},{y}) & = & \frac{{(\alpha x)}^T {y}}{\|{\alpha x}\| \|{y}\|} & = & \frac{\alpha {x}^T {y}}{|\alpha|\|{ x}\| \|{y}\|} & = & \frac{\alpha {x}^T {y}}{\alpha\|{ x}\| \|{y}\|} & = & \frac{ {x}^T {y}}{\|{ x}\| \|{y}\|}
\end{eqnarray*}
with $\alpha > 0$ as in our case. 

While for the translation:

\begin{eqnarray*}
\textrm{Cosine}({\beta + x},{y}) & = & \frac{{(\beta + x)}^T {y}}{\|{\beta + x}\| \|{y}\|} & \neq & \frac{ {x}^T {y}}{\|{ x}\| \|{y}\|}
\end{eqnarray*}

Correlation is both scale and translation invariant if the two vectors verse is the same. 

Conversely the Extended Jaccard Measure is clearly both scale and translation sensitive:
\begin{eqnarray*}
\textrm{ExtendedJaccard}({\alpha x},{y}) & = & \frac{{(\alpha x)}^T {y}}{\|{\alpha x}\|^2 + \|{y}\|^2 - {(\alpha x)}^T {y}} & = & \frac{\alpha{x}^T {y}}{\alpha^2\|{x}\|^2 + \|{y}\|^2 - \alpha{x}^T {y}} & \neq & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}}
\end{eqnarray*}
The same occurs for translation.
