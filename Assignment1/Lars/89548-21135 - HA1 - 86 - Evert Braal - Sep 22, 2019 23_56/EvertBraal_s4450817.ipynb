
1.2.2 Principal Component Analysis, more commonly known as PCA, can be used to visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

--------

_write your answer to 1.2.2 here_

With PCA, correlated variables will be transformed into a smaller number of uncorrelated variables. These variables are called principal components.
The first principal component accounts for as much of the variability in the data as possible, each succeeding component accounts for as much of the remaining variability as possible.
PCA should be performed on a square matrix.

In PCA the axes of the original variable coordinate system are more or less rotated to new orthogonal axes. So PCA is a dimensionality reduction method. With this reduction, data will be lost, 1 of the drawbacks, but there is less space neccesary so computations can be done faster too.

The limitations of PCA are:
The principal components should be a linear combinationof the original features. If not, the outcome isn't useful.
PCA assumes that the principle components are orthogonal.
When there are several variances, the higher variances will be treated as principal components, the lower ones will be treated as noise.

-------

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

------
_write your answer to 1.2.2 here_

Otherwise the data isn't normalized yet.
When there are variables with different units, like one is measured in thousands, so 10000, 20000 etc, and the other variable is measured in tens, so 10, 20 ... etc, then without mean center the data, pca will skip the second variable, since that one has less variation in comparison with the first variable.
So by mean centering the data, the data will be normalized so pca will give the best outcomes.

-------------

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute. Then you can subtract **μ** from **X** (**Y** = **X** - **μ**, where **μ** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
mean = X.mean(axis=0)
norm_data = X - mean

U, Sv, Vt = np.linalg.svd(norm_data)

V = Vt.T

Z = np.dot(norm_data,V)
plt.scatter(Z[:,0], Z[:,1])
plt.xlabel('Principle Component 1')
plt.ylabel('Principle Component 2')
plt.title('Principle Component Analysis of Nanonose')
plt.show()
print('Figure 2: Projection of the data onto the first two PCs')
```

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values. 
> e. (1 point) Calculate the row vector $\rho$ containing all values of $\rho$ for all PCs **m**. Create a bar plot with the variance explained on the Y axis and the number of the PC on the X axis. 

```python
rho = (Sv*Sv) / sum(Sv*Sv)
amountOfPCs = len(Z[0])

PCs = np.array([])
for i in range(amountOfPCs):
    PCs = np.append(PCs, i+1)

plt.bar(PCs, rho)
plt.xlabel('PCs')
plt.ylabel('Percentage variance explaned')
plt.title('Variance explained per PC in %')
plt.show()
print('Figure 3: Bar chart with every PC on X axis and the percentage of the variance that is explained on the Y axis')
```

> f. (0.5 points) How much of the variance is explained by the first three PCs?

```python
explained = (rho[0] + rho[1] + rho[2]) * 100
print('{0:.2f}% of the variance is explained by the first three PCs'.format(explained))
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

```python
checkCosineAlpha1 = similarity(x, y, 'Cosine')
checkCosineAlpha2 = similarity(alpha*x, y, 'Cosine')

if (checkCosineAlpha1.all() == checkCosineAlpha2.all()):
    print('First statement is true')
else:
    print('First statement is false')

checkExJacAlpha1 = similarity(x,y, 'ExtendedJaccard')
checkExJacAlpha2 = similarity(alpha*x, y, 'ExtendedJaccard')

if (checkExJacAlpha1.all() == checkExJacAlpha2.all()):
    print('Second statement is true')
else:
    print('Second statement is false')
    
checkCorAlpha1 = similarity(x,y, 'Correlation')
checkCorAlpha2 = similarity(alpha*x,y, 'Correlation')

if (checkCorAlpha1.all() == checkCorAlpha2.all()):
    print('Third statement is true')
else:
    print('Third statement is false')
    

checkCosineBeta1 = similarity(x, y, 'Cosine')
checkCosineBeta2 = similarity(beta+x, y, 'Cosine')

if (checkCosineBeta1.all() == checkCosineBeta2.all()):
    print('Fourth statement is true')
else:
    print('Fourth statement is false')

checkExJacBeta1 = similarity(x,y, 'ExtendedJaccard')
checkExJacBeta2 = similarity(beta+x, y, 'ExtendedJaccard')

if (checkExJacBeta1.all() == checkExJacBeta2.all()):
    print('Fifth statement is true')
else:
    print('Fifth statement is false')
    
checkCorBeta1 = similarity(x,y, 'Correlation')
checkCorBeta2 = similarity(beta+x,y, 'Correlation')

if (checkCorBeta1.all() == checkCorBeta2.all()):
    print('Sixth statement is true')
else:
    print('Sixth statement is false')
```

Only the ExtendedJaccard statements aren't true. To explain it, we've to take a look at the formula's.

Regarding Cosine: if for example the words are alpha times bigger, than the 2 documents are still the same. It doesn't matter if the matrix exists of all ones or all fives, both matrices exist of those ones or fives.

Regarding Correlation: If all values are changed with the same value, then the correlation matrix doesn't change, since the 'meaning' of the values aren't changed. And the standard deviation doesn't change either, because all values are changed with the same number, so the differences between the values are still the same. 


Regarding ExtendedJaccard, x will be squared in the denominator while in the nominator x is just x . So when x changes with alpha or beta, the denominator will change diffently (it will change with alpha^2) than in the nominator(will change with only alpha). Therefore, you'll get a different outcome.
Example: 5/5 = (2/2) * 5/5 but 5/5 != (2/2^2) * 5/5
