1.2.2 Principal Component Analysis, more commonly known as PCA, can be cleverly used to better visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

## H2 What Is PCA and How Is it Used?
PCA is a feature extraction algorithm, i.e. an algorithm for dimension reduction which keeps the information provided by those dimension that were reduced.
PCA is used to reduce the number of dimensions (variables) by focusing on the most impactful of them. Dimension reduction is needed to decrease the chance of overfitting, as it reduces the number of relationships between variables that has to be considered, and it also allows for aesier visualisation of the data.
PCA is used because if the less important variables (dimensions) are simply eliminated, we cannot derive any information from these variables. PCA, however, projects the data points into a new space with less dimensions, so even information from those variables that are less important is still preserved.
PCA is used when we want to reduce the number of dimension and are unable to chose a variable (dimension) to remove from the consideration, when we want the new variables to be independent of one another and when we do not mind the variables might be less interpretable.

## H2 The Algorithm
Simply put, PCA creates new variables (dimensions) which are a combination of the previous variables (dimensions) and eliminates those dimensions that do not provide much information. Since the new variables are a combination of the old ones, every new variable contains information on all the old variables. 
More in depth explanation of the algorithm follows:

1. Get a covariance matrix (a matrix summarising relations of all the variables)
Take the matrix X of the independent variables and substracting the mean of each column from each entry of that column, creating a matrix X'. Every column in X' has a mean of zero. 
The matrix X' can be standardized into matrix Z by dividing each value by its' column's standard deviation. Each column of Z has standard deviation of 1.
If the standardization did not take place, let Z = X'.
Covariance matrix of Z is Z<sup>T</sup>Z

2. Get eigenvectors and eigenvalues of Z<sup>T</sup>Z
This is done via eigendecomposition: Z<sup>T</sup>Z = PDP<sup>-1</sup>, where P is the matrix of eigenvectors and D is the diagonal matrix with eigenvalues.
The first element in D is the first eigenvalue and it corresponds to the first column of P (the first eigenvector).
In the case that the matrix is not square, this is done via SVD described in the next exercise.

3. Get a sorted matrix of eigenvectors P*
Sort D on eigenvalues from the highest to the lowest and sort P accordingly.

4. Get the resuting Z*
Z* = ZP*
Each data point in Z* is a combination of the original variables with the weights determined by the eigenvector.
Each column of Z* is independent of one another.

5. Drop certain dimensions
3 possibilities how to determine which dimensions to keep
a) select according to the indended use - e.g. if we want to present data graphically, 2D or 3D is the best, so we only keep the 2 or 3 most important dimensions
b) determine how much variance (in %) you want to explain and keep just enough dimensions to reach the threshold
c) add the dimensions from the most important to the least important to the result until a they do not contribute significantly to the cummulative variance explained

## H2 Limitations of PCA
1. Linearity
PCA assumes the principle components are linear combinations of the original features. If this is not true, the results might not be correct.
2. Importance is based on variance
The axes with high variance will always be treated as the most important while those with low variance will be treated as noise.
3. PCA assumes the principle components are orthogonal


To apply PCA we must first mean center the data. mean centering means that the mean value for an attribute is subtracted from that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

If we do not mean center the data, the dimension with highest values will be favoured instead of the dimension with highest variance. <br /><br />
To illustrate, let us suppose we have **two data points in a 2D space**:<br />
**[50,1] and [52,10]**<br />
The resulting uncentered matrix is **Z [[50,1],[52,10]]** and centered matrix is **Z' [[-1,-4],[1,5]]**<br />
**Z<sup>T</sup>Z is [[5204,570],[570,101]]<br />
Z'<sup>T</sup>Z' is [[2,9],[9,41]]** <br />
Seeing that for the first example, the highest element is on position 1,1 and for the first, it is on 2,2, we can reasonably predict for the first example, eigenvalue corresponding to the x axis will be higher but for the second example, eigenvector corresponding with the y axis will be higher.<br />
Since y axis shows higher variance, it should be prefered (be the first PC) but as a result of forgetting to mean center the variables, x axis is actually prefered by the algorithm.

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X**, must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute and then subtracting it from **X** (**Y** = **X** - **1μ**, where **1** is a column vector of size **N** x 1, with N indicating the number of observations in the original data).

Then, the Singular Value Decomposition, or SVD, of **Y**, the mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the numpy.linalg.svd() function.

Using SVD on **Y** yields a series of vectors which can be used to project the data onto specific Principal Components (PCs). The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,1]** would yield the projection of the data onto the first PC.

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

X = pd.read_excel('nanonose.xls', sheet_name='Sheet1', header=0, index_col=[0,1], usecols='A, B, D:K', skiprows=[1])
X.as_matrix()

# mean centering
u = np.array(X.mean(axis=0)).reshape(1,8)
n = X.shape[0]
N = np.ones((n,1))
Y = X - (np.dot(N,u))

# decomposition
U, s, VT = np.linalg.svd(Y)
V = VT.T

# reduction
Z = np.dot(Y,V[:,:2])

# graph
plt.scatter(Z[:,0],Z[:,1])
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title("PCA applied on the first two PC")
plt.show()
```

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

In EDV, **A = PDP<sup>-1</sup>** where A is a square matrix, P is the matrix of eigenvectors and D is the diagonal matrix of eigenvalues.<br />
The main difference is that EVD can be used only in a specific case - when A is **a square matrix** and SVD can be used on any matrix.<br /><br />
Other differences are:<br />
1. The vectors of P in EVD are not necessarily orthogonal while vectors of U and V from SVD are orthonormal<br />
this is important in rotations/flips as vectors can only represent rotations if they form an orthogonal basis<br />
2. U and V do not have to be inverses of one another, while P and P<sup>-1</sup> always are inverses of one another<br />
3. In SVD, values in s are always real and non-negative, while in D, values can be any complex numbers<br />


PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

X = pd.read_excel('nanonose.xls', sheet_name='Sheet1', header=0, index_col=[0,1], usecols='A, B, D:K', skiprows=[1])
X.as_matrix()

# mean centering
u = np.array(X.mean(axis=0)).reshape(1,8)
n = X.shape[0]
N = np.ones((n,1))
Y = X - (np.dot(N,u))

# decomposition
U, s, VT = np.linalg.svd(Y)
V = VT.T

# getting p
p = s**2/np.sum(s**2)

#graph
plt.bar(list(range(1,len(p)+1)),p)
plt.xlabel('PCs')
plt.ylabel('Variance Explained [%]')
plt.title("Variance explained by each PC")
plt.show()
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 gray scale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\end{eqnarray*}
where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary
data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary,
we will transform the data to be binary when calculating these two measures of
similarity by setting
$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$

```python
from Toolbox import similarity as s
import numpy as np

# arbitrary matrices and constants
X = np.array([[1,2,4],[6,8,4],[6,7,5]])
Y = np.array([[4,6,1],[7,5,2],[5,4,1]])
a = 9
b = 3

# similarity measures without any manipulations
cos = s.similarity(X, Y, 'cos')
ext = s.similarity(X, Y, 'ext')
cor = s.similarity(X, Y, 'cor')

# multiplication
cosA = s.similarity(np.dot(a,X), Y, 'cos')
extA = s.similarity(np.dot(a,X), Y, 'ext')
corA = s.similarity(np.dot(a,X), Y, 'cor')

# addition
cosB = s.similarity(b + X, Y, 'cos')
extB = s.similarity(b + X, Y, 'ext')
corB = s.similarity(b + X, Y, 'cor')

#comparisons
print("Cosine equals Cosine with multiplication:\n",np.allclose(cos,cosA))
print("Cosine equals Cosine with addition:\n",np.allclose(cos,cosB))

print("\nExtended Jaccard equals Extended Jaccard with multiplication:\n",np.allclose(ext,extA))
print("Extended Jaccard equals Extended Jaccard with addition:\n",np.allclose(ext,extB))

print("\nCorrelation equals Correlation with multiplication:\n",np.allclose(cor,corA))
print("Correlation equals Correlation with addition:\n",np.allclose(cor,corB))
```

Since Cosine compares the cosine between two vectors and does not take their length into the account, **the first formula holds**.<br />
