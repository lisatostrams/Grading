> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

Answer to question 1.2.2a

What is PCA? PCA = Principal Component Analysis. It is a dimension reduction method to capture the variability of the data, so to find the largest amount of variation in the data set. It's a linear algebra technique for continuous attributes that finds new attributes or principal components that firstly are linear combinations of the original attributes, secondly are perpendicular to each other and thirdly that capture the maximum amount of variation in the data. PAC can be used to see how similar lots of different data groups are. <br>
A PCA plot converts the correlations (or lack there of) among all of the cells into a graph. It finds these new attributes (principal components): the first dimension captures as much of the variability as possible, then the second captures the largest amount of variation after that one, etc. These principal components can be found by computing the eigenvalues as they correspond to the eigenvectors of the covariance matrix. <br>
The distances beteen the clusters you see in the PCA graph show the similarity between the clusters. On the x-axis are the horizontal distances between the clusters, which say 'more' than the vertical distances (y-axis). So if the distances between the clusters are about the same horizontally and vertically, then that still means that the two clusters are more different from each other than the ones with the vertical distance. <br>
PCA is very closely related to SVD, which stands for Single Value Decomposition. For SVD, the mean of the data isn't removed, while for PCS, it is.

When can PCA be used? PCA can be used in situations like pattern finding or dimensionality reduction. This is because PCA has some very attractive properties, so it can be used in any situation where these properties are important. 
Firstly PCA tends to identify the strongest patterns in the data. Therefore it's very attractive to use it for pattern-finding techniques. <br>
A second property of PCA is that PCA is very good at filtering out noise. This is because PCA takes only the most important parts of the data (noise isn't (shouldn't at least) be the most important part of the data, so it's removed from the dataset). This makes PCA very attractive to be used for data mining or other data analysis algorithms. 
Thirdly it captures the most important parts of the data. PCA captures most of the variability of the data by a small fraction of the total set of dimension. Therefore when PCA is applied to data, it becomes more low-dimensional, meaning this data can be used to apply techniques to that don't work well with high-dimensional data. 

What are drawbacks and limitations? One limitation is that a few dimensions will be dominating the PCA as the directions with the largest variance are assumed to be of most interest. This is because the original dimensions that intrinsically have a much larger variance will have most impact on the principal component. This can be prevented however by normalizing the data. <br>
Another limitation is that PCA will become useless if there's too much noice in the data. PCA filters out the noice, but if there's too much of it, it will base the principal components on noise instead of the pattern.

Answer to question 1.2.2b

We first need to mean center the data before applying PCA because it makes sure that the first PC does not only explain the mean of each variable but the variance. When mean-centering using SVD shifting the data makes sure that the center is on top of the origin in the graph. However this does not change the way the data points are positioned relative to each other. In the second option for PCA, EVD, so when using eigenvalues and vectors, mean-centering is done in order to calculate the covariance matrix. The eigenvalues correspond to the variance of the data along the eigenvectors after eigendecomposition is applied to the covariance matrix. 

-------------

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute. Then you can subtract **μ** from **X** (**Y** = **X** - **μ**, where **μ** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
##Answer to question 1.2.2c

# mean center the data
Y = X - X.mean(axis=0)
Y = np.matrix(Y, dtype='float')

# perform SVD
U,S,V = np.linalg.svd(Y)
V = V.transpose() #transpose V

# project the data onto the PCs
PC_1 = np.dot(X, V[:,0]) #PC1
PC_2 = np.dot(X, V[:,1]) #PC2

# scatter plots
plt.title('Visualisation of the projection of the data onto the first two principal components')
plt.xlabel('PC_1')
plt.ylabel('PC_2')
plt.scatter([PC_1], [PC_2])
plt.show()

#Elaboration: 
#To calculate PCA I first I centered the data (saved in Y), then used SVD to get U, S and V using numpy.lingalg.svd(). This 
# yiels the unitary matrix having left singular vectors as columns, U, the singular values, sorted in non-increasing order,
# S, and the unitary matrix having right singular vectors as rows, V. Only V is needed (though it has to be transposed). I 
# projected the data onto the PCs by multiplying X with the first column of V for the first componend and then with the 
# second column of V for the second principal component. 
```

Answer to question 1.2.2d

EVD = eigenvalue decomposition. SVD is a diagonal matrix of singular values, EVD is a diagonal matrix of eigenvalues. They cannot both be applied in each case (see difference 2). SVD and EVD are the same when the matrix as a positive, definite, symmetric matrix and when for the calculation of SVD mean-centering is used. 

The differences between SVD and EVD are: 
1. EVD only uses one basis, namely only eigenvectors, while SVD uses two bases, namely both the left and the right singular vectors
2. For SVD the eigenbasis is orthonormal, while for EVD the basis of the eigendecomposision is not necessarily orthogonal

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$
