
1.2.2 Principal Component Analysis, more commonly known as PCA, can be used to visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

--------
Principal Component Analysis is a dimensionality reduction technique; it reduces the amount of time and memory required by data mining algorithms (by reducing a large set of variables to a smaller one that still contains most of the information form the initial set) and allows data to be more easily visualized. It may help to eliminate irrelevant features or reduce noise.

When can it be used?
It is used to tackle the curse of dimensionality; when too many dimensions in your data set cause difficulties and noise. It can be used for variables with strong correlation. If the variables are weakly correlated, PCA does not work well (drawback). Another limitation is that PCA relies on linear assumptions; if you have some variables in your data set that are not linearly correlated, PCA does not suffice.

How does it work?
The goal of PCA is to find a projection that captures the largest amount of variation in the data (this projection makes the 'first principal component'). Then find the eigenvectors of the covariance matrix; these define the new space. The direction orthogonal to the principal direction with then the largest variation makes the second principal component etc.

-------

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

------
If you don't standardize before applying PCA, you will get misleading directions. Mean centering gives direction of variability across the mean of the samples. Without doing this, you would be looking at variations around the origin.
(You have to normalize data to make sure distances mean more or less the same/to make sure every attribute has the chance to play a role in your model! Otherwise certain features (e.g. weight in mg) will be dominant over others (e.g. length in meters))
(mean centering is also called 'mean substraction' or standardizing. Normalization is different from standardization.)
Is normalization also needed for PCA?

-------------

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute. Then you can subtract **** from **X** (**Y** = **X** - ****, where **** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
# 1. calculate the mean
m = data_nrs.mean(axis=0)
# 2. center the data
Y = data_nrs - m
# 3. Compute the singular values and vectors
U,sv,Vt = np.linalg.svd(Y) #yields 3 matrices
V = Vt.T
# 4. Calculate the lower dimension representation
m1 = np.dot(Y,V[:,0]) #first principal component
m2 = np.dot(Y,V[:,1]) #second principal component

plt.scatter(m1, m2)
plt.title('Data onto PC1 and PC2')
plt.xlabel('PC1')
plt.ylabel('PC2')
print('This scatterplot visualizes the projection of the data onto the first two principal components.')


# @ == np.dot only for 2D, after that it is different!
```

EVD and SVD are similar, except for:

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values. 
> e. (1 point) Calculate the row vector $\rho$ containing all values of $\rho$ for all PCs **m**. Create a bar plot with the variance explained on the Y axis and the number of the PC on the X axis. 

```python
v = np.square(sv)
w = sum(np.square(sv))
v = v / w
print(v) #so first principal component explains about 76%, second principal component +- 11% etc.

sns.set_style('darkgrid')
labels = [1, 2, 3, 4, 5, 6, 7, 8]
plt.bar(labels, v)
plt.xlabel('Number of the PC')
plt.ylabel('Variance')
plt.title('Variance of each principal component')
print('\nThe data in this bar plot show the variance of the respective principal component.')
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

```python
import numpy as np
from Toolbox.similarity import similarity

X = np.array([1,2,3]) 
Y = np.array([4,5,6]) 
a = 6
b = 3

sim1 = similarity(X, Y, 'cos')
print(sim1)
sim1a = similarity(a*X, Y, 'cos')
print(sim1a)

sim2 = similarity(X, Y, 'ext')
print(sim2)
sim2a = similarity(a*X, Y, 'ext')
print(sim2a)

sim3 = similarity(X, Y, 'cor')
print(sim3)
sim3a = similarity(a*X, Y, 'cor')
print(sim3a)

sim4 = similarity(b+X, Y, 'cos')
print(sim4)
sim4a = similarity(X, Y, 'cos')
print(sim4a)

sim5 = similarity(1+b+X, Y, 'ext')
print(sim5)
sim5a = similarity(X, Y, 'ext')
print(sim5a)

sim6 = similarity(b+X, Y, 'cor')
print(sim6)
sim6a = similarity(X, Y, 'cor')
print(sim6a)
```

---

Cosine(,) = Cosine(,) 
see sim1 and sim1a. See cos formula: the numerator will be a times bigger, and vectorlength(a * p) will be a times bigger too (take p = (1, 2), then vectorlength(a* p) = sqrt(1a^2 + 4a^2) = a * sqrt(5) = a * vectorlength(p). The a in the numerator and denominator will be cancelled out, so Cosine(,) = Cosine(,) is correct.

ExtendedJaccard(,) = ExtendedJaccard(,)
see sim2 and sim2a. This statement is incorrect. See above for given formula ExtendedJaccard: the numerator will be 'a' times as big for ExtendedJaccard(,) compared to ExtendedJaccard(,), but the denominator will not be simply 'a' times as big (since vectorlengths x and y are squared and then added). Therefore 'a' will not cancel out in numerator and denominator and thus the statement is incorrect.

Correlation(,) = Correlation(,)
see sim3 and sim3a. This statement is correct.The correlation will not change if one of the data sets is multiplied by a, because the entire data set is moved, so the correlation stays the same.

Cosine(+,) = Cosine(,)
see sim4 and sim4a. This statement is incorrect, respective results are 1.0 and 0.9746... . This can easily been seen when filling in the cos formula; the 'b' will not cancel out in the numerator and denominator.

ExtendedJaccard(+,) = ExtendedJaccard(,)
see sim5 and sim 5a. This statement is incorrect. Filling in the formula shows that 'b' will not cancel out in the numerator and denominator.

Correlation(+,) = Correlation(,)
see sim6 and sim6a. This statement is correct, both results are 1. The correlation will not change if one of the data sets is 'moved up' by b, because the entire data set is moved, so the correlation stays the same.
