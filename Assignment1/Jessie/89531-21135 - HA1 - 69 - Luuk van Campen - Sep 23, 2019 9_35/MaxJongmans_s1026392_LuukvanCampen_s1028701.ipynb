The idea behind PCA is finding a different view for data so it can be seperated better. It can be used if, for example, you have a two dimensional dataset that is diagonally spreading. It is then possible to do PCA by creating a new axis roughly in the same direction as the data is spreading while making sure that the spread of the data on that axis is as large as possible. This axis can then be taken as the new horizontal axis. It is then possible to also draw a new vertical axis through the horizontal axis. It will then be visible that the spread of the data across the vertical axis is much smaller than the spread of the data across the horizontal axis. By doing this, the data on the vertical axis has become "less usefull". If PCA is being applied, the data on the higher numbered axes will become less and less spread. This means that while doing PCA, the axes are ordered in terms of their significance. This can be very useful for doing dimension reduction, because you can remove the higher numbered axes because their data will not be important. A drawback of PCA might be that it considers high-variation dimensions very important and low-variation dimensions as not important. This might lead to problems when using PCA on a dataset that has incorrect measurements that have a very high variance. This might also mean that PCA assigns data different value than the researcher intended. If the researches thinks some low variance data is more important than some high variance data, PCA will not take this into account. Another huge drawback of PCA is that it can only be used in linear manifolds. It is also limited by the fact that that the principal components are always orthogonal, which does not work for every set of datapoints.


-------------

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

--------
The origin of the new coordinate system are the means of both attributes. By doing mean centering, you adjust every value in each attribute such that it's position is correct relative to the new center of the coordinate system. After that you can decorrelate the data by rotating the new axes and you will be able to find the covariance matrix.

-------

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute. Then you can subtract **** from **X** (**Y** = **X** - ****, where **** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
Y = X - X.mean(axis = 0)
U, S, Vt = np.linalg.svd(Y)
V = Vt.T
Z = np.dot(Y, V)

plt.scatter(Z[:,0], Z[:,1])
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title("Nanonose data projected onto the first two PCs")
plt.show()

```

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

---

Differences: 
- Every matrix has an SVD. A matrix can only have an EVD if it has the same number of columns and rows.
- All vectors in SVD are orthogonal. That does not necessarily hold for EVD.
- SVD uses different (orthonormal) left and right vectors as it's basis, while EVD uses only it's eigenvectors as a basis.

Similarities:
- Both SVD and EVD can be used for dimensionality reduction.
- SVD is essentially a generalization of EVD for non-square matrices.

So EVD can only be applied if we have a square matrix. SVD can always be applied.

---

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

In the case of comparing pictures, it is better to use Cosine similarity. This is mostly due to its low complexity, which comes in handy when computing similarity between large vectors.

```python
from Toolbox.similarity import similarity
X = np.array([6, 7, 8, 9, 10, 11, 12])
Y = np.array([8, 9, 10, 11, 12, 13, 14])

A = similarity(500*X, Y, 'Cosine')
print(A)
B = similarity(X, Y, 'Cosine')
print(B)
C = similarity(500*X, Y, 'ExtendedJaccard')
print(C)
D = similarity(X, Y, 'ExtendedJaccard')
print(D)
E = similarity(500*X, Y, 'Correlation')
print(E)
F = similarity(X, Y, 'Corelation')
print(F)
G = similarity(500+X, Y, 'Cosine')
print(G)
H = similarity(X, Y, 'Cosine')
print(H)
I = similarity(500+X, Y, 'ExtendedJaccard')
print(I)
J = similarity(X, Y, 'ExtendedJaccard')
print(J)
K = similarity(500+X, Y, 'Correlation')
print(K)
L = similarity(X, Y, 'Correlation')
print(L)
```

---

Cosine(ax, y) = Cosine(x,y) gives exactly the same result. This is because cosine computes similarity based on the similarity between directions of vectors. Multiplying a vector by a scalar only changes the lenght of the vector and not its direction, so the similarity stays the same.

ExtendedJaccard(ax,y) and ExtendedJaccard(x,y) are not equal. This is because when multiplying the vector with a constant, the covariance matrix changes and this causes the similarity to the original vector to change.

Correlation(ax,y) and Correlation(x,y) are equal. This is because the similarity between the two zscores will remain the same.

Cosine(b + x, y) and Cosine(x, y) are not equal. This is because adding a constant to the vector changes the direction of the vector, thereby changing the similarity between directions of vectors.

ExtendedJaccard(b+x,y) and ExtendedJaccard(x,y) are not equal. This is because the covariance matrix changes when adding a constant to the vector. This will cause the similarity to the original vector to change.

Correlation(b+x,y) and Correlation(x,y) are equal. This is because the similarity between the two zscores will remain the same.
