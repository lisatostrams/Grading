> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

--------

The goal of PCA is to find a new set of dimensions (attributes) that better
captures the variability of the data.The first dimension is
chosen to capture as much of the variability as possible.

PCA tends to identify the strongest patterns in the data, so PCA can be used as a pattern-finding
technique. Often most of the variability of the data can be captured
by a small fraction of the total set of dimensions. As a result, dimensionality
reduction using PCA can result in relatively low-dimensional data and it may
be possible to apply techniques that don't work well with high-dimensional
data. Since the noise in the data is (hopefully) weaker than the patterns,
dimensionality reduction can eliminate much of the noise. This is beneficial
both for data mining and other data analysis algorithms.

Tan, P. N., Steinbach, M., & Kumar, V. (2016). Introduction to data mining. Pearson Education India.


Limitations:

Linearity: PCA assumes that the principle components (PCs) are a linear.

Large variance implies more structure: In PCA  variance is used as the measure of how important a particular dimension is. High variance axes are more likely to be treated as PCs, while low variance axes are more likely to be treated as noise.

Orthogonality: PCA assumes that the principle components are orthogonal.

Drawback:

 time or space complexity of O(m2) or higher (where rm is
the number of objects), and thus, are not practical for large data sets.


-------

To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute (i.e. a column) is subtracted from all values for that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

mean centering is necessary for PCA to ensure that the first PC describes the direction of maximum variance. If mean subtraction is not performed, the first PC might instead correspond to the mean of the data and may pierce the cloud not along the main direction of the cloud, and will be misleading(for statistics purposes).

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X** must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute. Then you can subtract **μ** from **X** (**Y** = **X** - **μ**, where **μ** is subtracted from every row.).

Then, the Singular Value Decomposition (or SVD) of **Y**, the now mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the `numpy.linalg.svd()` function.

Using SVD on **Y** yields 3 matrices **U**, **S** and **V**<sup>T</sup>. These are used to project the data onto specific Principal Components (PCs). 

The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data, and **Z[:,0]** indicates the data projected onto the first PC. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,0]** would also yield the projection of the data onto the first PC.

_Note that for matrices, the notation **Y\*V** denotes the `dot` product between **Y** and **V**!_

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
import numpy as np

def computeMean(x):
    column=x.shape[0]
    row=x.shape[1]
    sum=np.zeros(shape=(row,1))
    mean=np.zeros(shape=(row,1))
    for j in range(row):
        for i in range(column):
           sum[j]+=x[i][j]
        mean[j]=sum[j]/column
    return mean

def computeMeanCentered(x,u):
    column=x.shape[0]
    row=x.shape[1]
    sum=np.zeros(shape=(column,row))
    for j in range(row):
        for i in range(column):
           sum[i][j]=x[i][j]-u[j]
    return sum
    
u=computeMean(x)

y=computeMeanCentered(x,u)
plt.scatter(y[:,0],y[:,1])
plt.xlabel("Sensor A")
plt.ylabel("Sensor B")
plt.title('MeanCentered sensorvalues')

plt.show()
U, S, Vt=np.linalg.svd(y)
V = Vt.T
A = np.dot(y,V)
plt.scatter(A[:,0],A[:,1])
plt.xlabel("First PC")
plt.ylabel("Second PC")
plt.title('Principal component analysis')
plt.show()
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 grayscale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\\
\end{eqnarray*}
<br>where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary, the script will transform the data to be binary when calculating these two measures of similarity by setting

$$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$$

You can generate these randomly or set them as you see fit. Using Python, calculate the following similarity measures, and check if the statements below are correct.

\textrm{Correlation}(\beta + {x},{y}) & = & \textrm{Correlation}({x},{y})

```python
from similarity import *

X=np.array([[2, 0, 1, 1, 0, 2, 1, 1]])
Y=np.array([[2, 1, 1, 0, 1, 1, 1, 1]])

def computeTranslationAlpha(X,Y,method,scaling):
    X=X*scaling
    print("Method=",method,"scaling=",scaling,similarity(X,Y , method))
    
def computeTranslationBeta(X,Y,method,scaling):
    X=X+scaling
    print("Method=",method,"scaling=",scaling,similarity(X,Y , method))
    
alpha1=1
alpha2=2
beta1=0
beta2=1

computeTranslationAlpha(X,Y,'cos',alpha1)
computeTranslationAlpha(X,Y,'cos',alpha2)
computeTranslationAlpha(X,Y,'ext',alpha1)
computeTranslationAlpha(X,Y,'ext',alpha2)
computeTranslationAlpha(X,Y,'cor',alpha1)
computeTranslationAlpha(X,Y,'cor',alpha2)

computeTranslationBeta(X,Y,'cos',beta1)
computeTranslationBeta(X,Y,'cos',beta2)
computeTranslationBeta(X,Y,'ext',beta1)
computeTranslationBeta(X,Y,'ext',beta2)
computeTranslationBeta(X,Y,'cor',beta1)
computeTranslationBeta(X,Y,'cor',beta2)
```

The functions cos and cor keep the same with a different constant alpha, while extended jaccard changes. When we look at the formula of cosine similiarity there is a x in the nominator and in the denominator, hence these cancel out. The same goes for correlation.

When the constant  is increased, the similairity value computed by ext jaccard decreases. This is in line with the formula written above, there are two instances of x in the denominator and only 1 instance of x in the nominator, hence the value decreases when the x is bigger.

For the beta factor both the value computed by the cos and extended jaccard method changes, but the value computed by the correlation method stays the same.

in conclusion:
The statements of cos and cor hold when there is a constant alpha. The same applies for cor when there is a beta constant.
