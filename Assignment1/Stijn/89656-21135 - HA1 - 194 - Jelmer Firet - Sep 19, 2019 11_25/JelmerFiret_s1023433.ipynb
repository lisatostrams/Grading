
1.2.2 Principal Component Analysis, more commonly known as PCA, can be cleverly used to better visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

PCA is a method to find which set of orthogonal axes best describe a dataset. These axes are ordered based on how effective they are. This information can be used to determine how to lower the dimensionality of a dataset that disrupts the data retrieval the leasts.

To apply PCA we must first mean center the data. mean centering means that the mean value for an attribute is subtracted from that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

The PCA algorithm finds the direction trought the origin that best represents the data; If the data were to be off-center than this would probably be the direction to the data cluster, instead of a direction within the data cluster. To be able to find the second direction we mean-center the data so that the first direction won't be found.

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X**, must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute and then subtracting it from **X** (**Y** = **X** - **1**, where **1** is a column vector of size **N** x 1, with N indicating the number of observations in the original data).

Then, the Singular Value Decomposition, or SVD, of **Y**, the mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the numpy.linalg.svd() function.

Using SVD on **Y** yields a series of vectors which can be used to project the data onto specific Principal Components (PCs). The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,1]** would yield the projection of the data onto the first PC.

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
mu = np.mean(excel_np_array, axis=0)
nanonose_centered = excel_np_array-mu
U,S,nanonose_pca = np.linalg.svd(nanonose_centered)
nanonose_projected = nanonose_centered.dot(nanonose_pca.T)

plt.scatter(nanonose_projected[:,1],nanonose_projected[:,0])
plt.title('A plot of the nanonose data along it\'s first two principal axes')
plt.xlabel('First principal axis')
plt.ylabel('Second principal axis')
print('This figure shows the data of the nanonose sensors, along it\'s first two principal axes')
```

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

SVD describes a transformation as a rotation, then a squash and then again a rotation.  
EVD describes a transformation as an invertable transformation, then a squash and then the inverse of the first transformation.  
Both try to find axes along which to perform a squash: the principle axes, but with SVD those are guaranteed to be orthogonal whereas for EVD this doesn't have to be the case.
EVD can't always be applied, it has to be performed on a square matrix. To solve this EVD is applied on the covariance matrix and even then the covariance matrix computed of the data might not be [diagonolizable](https://en.wikipedia.org/wiki/Diagonalizable_matrix).  
SVD on the other hand can always be applied

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values. 
> e. (1 point) Calculate the row vector $\rho$ containing all values of $\rho$ for all PCs **m**. Create a bar plot with the variance explained on the Y axis and the number of the PC on the X axis. 

```python
def rho(m,S):
    sum = 0.0
    for s in S:
        sum += s*s
    return S[m]*S[m]/sum

rhos = np.array([rho(m,S) for m in range(0,8)])

plt.bar(range(1,9),rhos)
plt.xlabel("Principal axis inde")
plt.ylabel("Variance explained by principal axis")
plt.title("Variance of principle axes")
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 gray scale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\end{eqnarray*}
where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary
data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary,
we will transform the data to be binary when calculating these two measures of
similarity by setting
$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$

> Even though a similarity measure is theoretically invariant e.g.\ to scaling, it might not be exactly invariant numerically.

```python
from Toolbox.similarity import similarity
x = np.array([[.492,.510,.242,.625],[.894,.410,.915,.592]])
y = np.array([[.762,.638,.921,.337],[.311,.136,.843,.541]])
alpha = 2.0
beta = .930
print ("Scaling input of Cosine similarity measure is invariant:")
print(similarity(x,y,'cos'))
print(similarity(x*alpha,y,'cos'))
print ("Scaling input of Extended Jaccard similarity measure is not (numerically) invariant:")
print(similarity(x,y,'ext'))
print(similarity(x*alpha,y,'ext'))
print ("Scaling input of Correlation similarity measure is invariant:")
print(similarity(x,y,'cor'))
print(similarity(x*alpha,y,'cor'))
print("")
print ("Shifting input of Cosine similarity measure is not (numerically) invariant:")
print(similarity(x,y,'cos'))
print(similarity(x+beta,y,'cos'))
print ("Shifting input of Extended Jaccard similarity measure is not (numerically) invariant:")
print(similarity(x,y,'ext'))
print(similarity(x+beta,y,'ext'))
print ("Shifting input of Correlation similarity measure is invariant:")
print(similarity(x,y,'cor'))
print(similarity(x+beta,y,'cor'))
```

The cosine similarity is invariant to scaling; It calculates the cosine of the angle between the two vectors, and this angle doesn't change if one of the vectors is scaled.
The Extended Jaccard similarity should be invariant to scaling, but because the exponential function is used a lot of accuracy get's lost.
The correlation similarity is invariant to scaling, as the covariance is divided by the variance of the axis, cancelling any scaling applied.

The cosine similarity is not invariant to shifting. It moves the data to another location, changing the angles.
I'm not entirely sure how extended Jaccard works, but I suspect it isn't invariant to shifting
The correlation similarity is invariant to shifting, because in the calculation of the covariance and variance the data is adjusted such that the mean is 0.
